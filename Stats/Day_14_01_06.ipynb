## 🔰One Sample Varience🔰

```python
df = pd.read_excel('/content/drive/MyDrive/CDAC/CDAC_DataBook.xlsx' , sheet_name = 'faithful')
```

```python
erupt = df.eruptions
erupt.head(2)
```

```python
np.var(erupt,ddof = 1)
```

```python
# Ho varience (erupt) < 1.1

# H                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                            o varience (erupt) <= 1.1

# Ho varience (erupt) > 1.1 right tail test

# TS = (n-1) * varience/(claim value)

#      DDOF
ts = (len(erupt)-1) * np.var(erupt, ddof=1)/1.1
ts
```

```python
from scipy.stats import chi2
```

```python
p_value = 1 - chi2.cdf(ts, len(erupt)-1)
p_value
```

```python
chi2.sf(ts, len(erupt)-1)
```

```python
samp1 = erupt[30:70]
samp2 = erupt[130:180]

samp1.head(3),samp2.head(3)
```

```python
np.var(samp1,ddof = 1 )/np.var(samp2, ddof = 1)
```

```python
# Ho: var(samp1)/var(samp2) >= 1.4
# Ha: var(samp1)/var(samp2) < 1.4

samp_ratio = np.var(samp1,ddof = 1 )/np.var(samp2, ddof = 1)

ts  = samp_ratio/1.4
ts
```

```python
from scipy.stats import f
```

```python
f.cdf(ts, len(samp1)-1, len(samp2)-1)
```
