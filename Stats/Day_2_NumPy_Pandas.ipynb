```python
a1 = pd.DataFrame(([1,2,3],[12,13,14],[23,24,25]),index=(["A","B","C"]))
a1
```

```python
mydata=[
    ["dev",29,2003],
    ["arv",14,2004],
    ["sak",13,2003],
    ["bha",27,2003]]

my_data = pd.DataFrame( mydata ,index=[row[0] for row in mydata], columns=["Name","Date","Year"])
print(my_data)

print("------------------------------------------------------------------")
for i in mydata:
    print(f'"{i[0]}"',end=",")
print()
print("------------------------------------------------------------------")
print(','.join(f'"{row[0]}"' for row in mydata))
print("------------------------------------------------------------------")
my_data.reset_index()
print()
print(my_data)
```
