## Broadcasting

```python
# broacasting make it possible that to repeat the same operation across all the values
a1 = np.array([2,3,4,5,6])
a2 = 2*a1 + 5

a1      ## array([2, 3, 4, 5, 6])
a2      ## array([9, 11, 13, 15, 17])
a1 + a2 ## array([11, 14, 17, 20, 23])

a3 = np.array([1,2,3]) #operands could not be broadcast together with shapes (5,) (3,)
# a1+a3                  # recycling of aaray would not happen like R lang
```
