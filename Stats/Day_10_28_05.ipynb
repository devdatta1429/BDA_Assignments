```python
df  = pd.read_excel('/content/drive/MyDrive/CDAC_DataBook.xlsx' , sheet_name = 'faithful')
```

```python
erupt = df.eruptions

xbar = np.mean(erupt)
print(xbar)

n = len(erupt)
print(n)

sd = np.std(erupt , ddof =1)
print(sd)
```

1-sample z test/ 1 sample t test

Used to check whether the difference betweent the sample mean and claim value can be ignored or not.\
Used to check whether the difference betweent the sample mean and claim value is statistically significant or not

```python
# one sample z-test

ssw.ztest(erupt, value = 3.6 , alternative = 'smaller')
# (np.float64(-1.621495827018525), np.float64(0.052455672107759366))
#         test statistics       ,           p-value

ssw.ztest(erupt, value = 3.7 , alternative = 'smaller')
# (np.float64(-3.06646147570652), np.float64(0.0010830435959524984))
```
