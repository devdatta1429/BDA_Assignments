# ***Chi-Square***

*🔰Chi-Square for goodness fo fit🔰*

```python
counts = np.array([50,20,45,25])
pattern = np.array([0.4 , 0.1 , 0.3 , 0.2])
```

```python
exp_counts = sum(counts) * pattern
```

```python
exp_counts
```

```python
from scipy.stats import chisquare
```

```python
scipy.stats.chisquare(counts, f_exp=exp_counts)
```
