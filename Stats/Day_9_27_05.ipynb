```python
#                 z = (1 - 0.95)/2 = 0.025
z = np.abs( norm.ppf( (1- cl)/2))   # norm.ppf(0.025) ------> np.abs make everything positive
z
```

```python
#                 t = (1 - 0.95)/2 = 0.025          # for pythom we use the student't distribution rather than norm
t = np.abs(t.ppf((1-cl)/2,n-1) ) # t.ppf(0.025) ------> np.abs make everything positive , ddof = (n-1) degree of freedom
t
```

```python
# use 't' beacuse it will work on both small and big sample size 30. z is used for greater than 30(Normal)
low_est = xbar - t*stdev/n**0.5
low_est
```

```python
upper_est = xbar + t*stdev/n**0.5
upper_est
```

```python
size = upper_est - low_est
size
```
