# 🔰Box Plot🔰(whisker plot)

# Box Plot: A Visual Summary of Data

A **box plot** (or **box-and-whisker plot**) is a graphical representation of a dataset using a **five-number summary**. It helps visualize:

* The center of the data
* The spread of the data
* The presence of outliers
* The overall distribution

---

# 1. Key Formulas & Definitions

## Interquartile Range (IQR)

The **Interquartile Range (IQR)** measures the spread of the middle 50% of the data.

\mathrm{IQR}=Q_3-Q_1

Where:

* (Q_1) = First Quartile
* (Q_3) = Third Quartile

---

## Lower Limit (Lower Whisker Boundary)

Used to identify lower outliers.

\mathrm{Lower\ Limit}=Q_1-1.5\times \mathrm{IQR}

---

## Upper Limit (Upper Whisker Boundary)

Used to identify upper outliers.

\mathrm{Upper\ Limit}=Q_3+1.5\times \mathrm{IQR}

---

## Outliers

Any data point that lies:

* Below the **Lower Limit**
* Above the **Upper Limit**

is considered an **outlier**.

---

# 2. The Five-Number Summary

A box plot is built using the following five values:

| Statistic                  | Description                                        |
| -------------------------- | -------------------------------------------------- |
| **Minimum**                | Smallest value in the dataset (excluding outliers) |
| **First Quartile ((Q_1))** | Median of the lower half of the data               |
| **Median ((Q_2))**         | Middle value of the ordered dataset                |
| **Third Quartile ((Q_3))** | Median of the upper half of the data               |
| **Maximum**                | Largest value in the dataset (excluding outliers)  |

---

# 3. How to Build a Box Plot

## Step 1: Arrange the Data

Sort the dataset in ascending order.

Example:

```python
data = [2, 4, 5, 7, 8, 10, 12, 15, 18]
```

---

## Step 2: Find the Median ((Q_2))

The median is the middle value of the ordered dataset.

---

## Step 3: Find (Q_1) and (Q_3)

* (Q_1): Median of the lower half
* (Q_3): Median of the upper half

---

## Step 4: Calculate the IQR

Use:

[
\mathrm{IQR} = Q_3 - Q_1
]

---

## Step 5: Determine Outlier Limits

Calculate:

[
\mathrm{Lower\ Limit} = Q_1 - 1.5 \times \mathrm{IQR}
]

[
\mathrm{Upper\ Limit} = Q_3 + 1.5 \times \mathrm{IQR}
]

Any points outside these limits are outliers.

---

## Step 6: Draw the Box Plot

* Draw a box from (Q_1) to (Q_3)
* Draw a line inside the box for the median
* Extend whiskers to the minimum and maximum non-outlier values
* Plot outliers as individual points

---

# Example

Dataset:

```python
data = [5, 7, 8, 9, 10, 12, 15, 18, 22, 30]
```

Suppose:

* (Q_1 = 8)
* (Q_3 = 18)

Then:

[
\mathrm{IQR} = 18 - 8 = 10
]

Lower Limit:

[
8 - 1.5(10) = -7
]

Upper Limit:

[
18 + 1.5(10) = 33
]

Since all values lie between (-7) and (33), there are **no outliers**.

---

# Why Box Plots Are Useful

Box plots help you:

* Detect outliers quickly
* Compare distributions between datasets
* Understand spread and variability
* Identify skewness in data

They are widely used in:

* Statistics
* Data Science
* Machine Learning
* Exploratory Data Analysis (EDA)
