# 🚀 Ultimate Data Science & Machine Learning Study Notes

## 1. The Machine Learning Landscape

Machine Learning algorithms are broadly split into two main branches based on the structure of the input data:

### A. Supervised Learning
*   **Definition:** The algorithm learns from labeled training data (data with a known "correct answer").
*   **Goal:** Learn a mapping function from input variables (\(X\)) to an output variable (\(y\)).
*   **Types:**
    *   **Regression:** Predicts continuous numbers (e.g., predicting a house price or exam score).
    *   **Classification:** Predicts discrete categories (e.g., predicting if an email is "Spam" or "Not Spam").

### B. Unsupervised Learning
*   **Definition:** The algorithm analyzes unlabeled datasets to discover hidden structures and patterns without human intervention.
*   **Core Categories:**
    1.  **Clustering:** Groups similar data points together (e.g., **K-Means**, **Hierarchical Clustering**, **DBSCAN**).
    2.  **Dimensionality Reduction:** Compresses data features while preserving vital information (e.g., **PCA**, **t-SNE**).
    3.  **Association Rule Learning:** Finds hidden "if-then" relationships in transaction logs (e.g., **Apriori**, **FP-Growth**).
    4.  **Anomaly Detection:** Automatically flags unusual data spikes or outliers (e.g., **Isolation Forest**, **One-Class SVM**).

---

## 2. Exploring Data Relationships with Correlation

Before building a model, data scientists use **correlation** to measure how variables interact.

### Pandas Correlation: `df.corr(method='pearson')`
*   Measures the **Pearson Correlation Coefficient** between all numerical features.
*   The output is a square matrix showing scores ranging from **-1 to +1**.

### Key Statistical Boundary Scales:
*   **\(+1.00\) (Perfect Positive Correlation):** As variable \(X\) goes up, variable \(Y\) goes up in a perfect straight line.
*   **\(0.00\) (No Linear Correlation):** The variables have an absolute random dispersion; they share no linear trend.
*   **\(-1.00\) (Perfect Negative Correlation):** As variable \(X\) goes up, variable \(Y\) goes down in a perfect straight line.

---

## 3. Demystifying Linear Regression Under the Hood

Linear Regression models a continuous relationship between an independent feature (\(X\)) and a dependent target (\(y\)) using a straight-line algebraic equation:

\[\mathbf{y = mx + c}\]

*(In Machine Learning, this formula is also commonly written as \(y = wX + b\), where \(w\) is the weight and \(b\) is the bias).*

### What happens during execution?

```python
model = LinearRegression() # Step 1: Instantiates a blank mathematical structure.
model.fit(X, y)            # Step 2: Executes the training optimization phase.
```

1.  **Measures Residuals:** The model draws a random baseline and measures the vertical distances from every real observation point to the line. These distances are called **residuals (errors)**.
2.  **Ordinary Least Squares (OLS):** The algorithm squares all individual residuals and aggregates them. It leverages calculus-based optimization to find the exact slope and intercept that yield the lowest possible sum of squared errors.
3.  **Locks Parameters:** Once optimal results are found, the values are stored inside the model object as **`model.coef_`** (\(m\)) and **`model.intercept_`** (\(c\)).

---

## 4. What is a Coefficient?

In statistics and ML, a **coefficient** is a calculated number that quantifies the trend, strength, or volatility of a dataset.

### A. Machine Learning Coefficient (`model.coef_`)
*   It represents the **slope (\(m\))** of the regression line.
*   **Core Meaning:** It defines how many units \(y\) will change for every single 1-unit increase in \(X\).
*   *Example:* In the formula \(\text{Score} = (10 \times \text{Hours}) + 10\), the coefficient is **\(10\)**. Each extra hour adds exactly 10 score points.

### B. Core Statistical Coefficients
1.  **Regression Coefficient (\(m\)):** The directional slope and scale of impact of an independent feature.
2.  **Correlation Coefficient (Pearson's \(r\)):** Unit-free measurement tracking relationship strength and direction (\(-1\) to \(+1\)).
3.  **Coefficient of Determination (\(R^2\)):** A model performance accuracy metric scoring from \(0\) to \(1\). It details the percentage of real-world variance accounted for by the model.
4.  **Coefficient of Variation (\(CV\)):** Calculated as standard deviation divided by the mean (\(\sigma / \mu\)). Used to compare variance across different datasets.

---

## 5. Model Evaluation: Regression vs. Classification

Do not mix up evaluation frameworks! You cannot assess a numerical prediction line using a categorical sorting tool.

| Task Category | Output Structure | Right Metrics to Evaluate | Wrong Metrics (Do Not Use!) |
| :--- | :--- | :--- | :--- |
| **Regression** | Continuous Numbers (e.g., Scores, Prices, Temperature) | **\(R^2\) Score**, Mean Squared Error (**MSE**), Mean Absolute Error (**MAE**) | Confusion Matrix, Accuracy %, F1-Score |
| **Classification** | Categorical Buckets (e.g., Pass/Fail, Red/Blue/Green) | **Confusion Matrix**, Accuracy %, Precision, Recall, F1-Score | Mean Squared Error (**MSE**), \(R^2\) Score |

P(Y=1)=\frac{1}{1+e^{-z}}
$$

This is called the Sigmoid Function.

# 21. Sigmoid Function

Converts:

```text
-∞ to +∞
```

into:

```text
0 to 1
```

Example:

```text
Input = 3
Output = 0.95
```

# 22. Decision Boundary

Usually:

```python
if probability >= 0.5:
    class = 1
else:
    class = 0
```

Example:

```text
0.85 → Pass
```

# 23. Confusion Matrix

Used ONLY for Classification Problems.

Example:

|            | Predicted No | Predicted Yes |
| ---------- | ------------ | ------------- |
| Actual No  | TN           | FP            |
| Actual Yes | FN           | TP            |

---

## Terms

### TP

True Positive

Predicted Yes and Actually Yes

---

### TN

True Negative

Predicted No and Actually No

---

### FP

False Positive

Predicted Yes but Actually No

---

### FN

False Negative

Predicted No but Actually Yes

---

# 24. Accuracy

Measures overall correctness.

$$
Accuracy=
\frac{TP+TN}
{TP+TN+FP+FN}
$$

Example:

```text
90 correct out of 100
```

Accuracy:

```text
90%
```
