# Sigmoid Function (Logistic Sigmoid)

The **sigmoid function** (specifically the logistic sigmoid function) is one of the most famous curves in mathematics, machine learning, and deep learning.

It takes any real-valued number and maps it into a strict range between **0 and 1**. Because of this, it is incredibly useful for models that need to predict probabilities.

---

## 1. The Mathematical Formula

The standard logistic sigmoid function is defined as:

$$
S(x) = \frac{1}{1 + e^{-x}}
$$

Where:

* $x$ is the input value.
* $e$ is Euler's number (approximately 2.718).

---

## 2. Key Properties of the Curve

* **The "S" Shape:** The function creates an S-shaped curve.
* **Boundaries:** As $x$ goes to positive infinity, $S(x)$ approaches **1**. As $x$ goes to negative infinity, $S(x)$ approaches **0**.
* **The Midpoint:** When the input $x = 0$, the output is exactly **0.5**.
* **Monotonicity:** It is strictly increasing, meaning higher inputs always yield higher outputs.

---

## 3. Where Is It Used?

### Logistic Regression

In binary classification problems (e.g., predicting whether an email is **Spam** or **Not Spam**), logistic regression uses the sigmoid function to convert the model's raw score into a probability value between **0** and **1**.

$$
P(y=1 \mid x) = \frac{1}{1 + e^{-z}}
$$

where $z$ is the linear combination of input features.

A common decision rule is:

* If the output $\geq 0.5$, classify as **Spam**.
* If the output $< 0.5$, classify as **Not Spam**.

### Neural Networks (Activation Function)

Historically, the sigmoid function was a popular activation function because it introduces **non-linearity** into neural networks, enabling them to learn complex patterns.

The output of each neuron is squashed into the range $(0, 1)$, making it useful when the model needs probabilistic outputs.

---

## 4. The Main Drawbacks

Although elegant, the sigmoid function is rarely used in hidden layers of modern deep neural networks due to the following limitations:

### 1. Vanishing Gradient Problem

When the input $x$ becomes very large or very small, the sigmoid curve becomes almost flat.

Its derivative is:

$$
S'(x) = S(x)\bigl(1 - S(x)\bigr)
$$

Since the derivative approaches **0** at the extremes, gradients become extremely small during backpropagation, slowing or even preventing learning.

### 2. Not Zero-Centered

The sigmoid function always produces positive outputs:

$$
0 < S(x) < 1
$$

Because activations are not centered around zero, optimization can become inefficient and may cause gradients to zig-zag during training.

---

## 5. Modern Usage

Today, activation functions such as:

* **ReLU (Rectified Linear Unit)**
* **Leaky ReLU**
* **GELU**
* **Swish**

are generally preferred for hidden layers.

However, the **sigmoid function remains the standard choice for the final output layer of binary classification models**, where the output is interpreted as a probability.

---

## Summary

The sigmoid function is a smooth, S-shaped mathematical function that maps any real number to a value between **0 and 1**.

$$
S(x) = \frac{1}{1 + e^{-x}}
$$

### Advantages

* Produces probability-like outputs.
* Smooth and differentiable.
* Easy to interpret.

### Disadvantages

* Suffers from the vanishing gradient problem.
* Not zero-centered.
* Slower convergence compared to modern activation functions.

**Current Best Practice:** Use **ReLU-based activations** in hidden layers and **sigmoid** in the output layer for binary classification tasks.

# Tanh Function (Hyperbolic Tangent)

The **Tanh (Hyperbolic Tangent)** function is a popular activation function used in neural networks. Like the sigmoid function, it introduces non-linearity into a model, but it has the advantage of being **zero-centered**.

---

## 1. The Mathematical Formula

The tanh function is defined as:

$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

Where:

* $x$ is the input value.
* $e$ is Euler's number (approximately 2.718).

---

## 2. Key Properties of the Curve

* **S-Shaped Curve:** Similar to the sigmoid function.
* **Output Range:** Values lie between **-1 and 1**.

$$
-1 < \tanh(x) < 1
$$

* **Zero-Centered:** Outputs can be negative, zero, or positive.
* **Monotonic:** The function is strictly increasing.

---

## 3. Important Values

When:

$$
x = 0
$$

then

$$
\tanh(0) = 0
$$

As:

$$
x \rightarrow +\infty
$$

$$
\tanh(x) \rightarrow 1
$$

As:

$$
x \rightarrow -\infty
$$

$$
\tanh(x) \rightarrow -1
$$

---

## 4. Relationship with Sigmoid

The tanh function can be expressed using the sigmoid function:

$$
\tanh(x) = 2 \cdot \sigma(x) - 1
$$

where

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

is the sigmoid function.

---

## 5. Derivative of Tanh

The derivative of tanh is:

$$
\frac{d}{dx}\tanh(x) = 1 - \tanh^2(x)
$$

or equivalently,

$$
\tanh'(x) = 1 - (\tanh(x))^2
$$

This derivative is easy to compute and is useful during backpropagation.

---

## 6. Where Is It Used?

### Neural Networks

The tanh function is commonly used as an activation function in:

* Feedforward Neural Networks
* Recurrent Neural Networks (RNNs)
* Long Short-Term Memory Networks (LSTMs)

Since its outputs are centered around zero, optimization is often more efficient than with the sigmoid function.

---

## 7. Advantages

* Zero-centered outputs.
* Stronger gradients near zero than sigmoid.
* Smooth and differentiable.
* Produces both positive and negative activations.

---

## 8. Disadvantages

### Vanishing Gradient Problem

For very large positive or negative inputs, tanh saturates:

$$
\tanh(x) \approx 1
$$

or

$$
\tanh(x) \approx -1
$$

As a result,

$$
\tanh'(x) \approx 0
$$

which can cause gradients to vanish during training.

---

## 9. Comparison with Sigmoid

| Feature            | Sigmoid | Tanh                |
| ------------------ | ------- | ------------------- |
| Output Range       | $(0,1)$ | $(-1,1)$            |
| Zero-Centered      | No      | Yes                 |
| Saturation         | Yes     | Yes                 |
| Vanishing Gradient | Yes     | Yes                 |
| Hidden Layer Usage | Rare    | Better than Sigmoid |

---

## Summary

The **Tanh Function** is a zero-centered activation function that maps real-valued inputs into the range:

$$
-1 < \tanh(x) < 1
$$

and is defined by:

$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

It generally performs better than the sigmoid function in hidden layers because its outputs are centered around zero. However, it still suffers from the vanishing gradient problem, which is why modern deep networks often prefer **ReLU** and its variants.


# ReLU Function (Rectified Linear Unit)

The **ReLU (Rectified Linear Unit)** is the most widely used activation function in modern deep learning. It introduces non-linearity while being computationally simple and efficient.

---

## 1. The Mathematical Formula

The ReLU function is defined as:

$$
\text{ReLU}(x) = \max(0, x)
$$

Alternatively, it can be written as:

$$
\text{ReLU}(x) =
\begin{cases}
0, & x < 0 \
x, & x \geq 0
\end{cases}
$$

Where:

* $x$ is the input value.

---

## 2. Key Properties of the Curve

* Outputs **0** for all negative inputs.
* Outputs the input itself for positive inputs.
* Computationally efficient.
* Introduces non-linearity into neural networks.

---

## 3. Important Values

When:

$$
x = -5
$$

$$
\text{ReLU}(-5) = 0
$$

When:

$$
x = 0
$$

$$
\text{ReLU}(0) = 0
$$

When:

$$
x = 5
$$

$$
\text{ReLU}(5) = 5
$$

---

## 4. Output Range

The output range of ReLU is:

$$
0 \leq \text{ReLU}(x) < \infty
$$

---

## 5. Derivative of ReLU

The derivative is:

$$
\frac{d}{dx}\text{ReLU}(x)
$$
==========================
$$
\begin{cases}
0, & x < 0 \
1, & x > 0
\end{cases}
$$

At:

$$
x = 0
$$

the derivative is technically undefined, but in practice deep learning frameworks assign a value of either **0** or **1**.

---

## 6. Why ReLU Is Popular

Unlike Sigmoid and Tanh, ReLU does not saturate for positive inputs.

For:

$$
x > 0
$$

$$
\text{ReLU}'(x) = 1
$$

This helps gradients flow through deep networks and significantly reduces the **vanishing gradient problem**.

---

## 7. Advantages

* Computationally simple.
* Faster training.
* Reduces vanishing gradients.
* Sparse activation (many neurons output zero).
* Works well in deep neural networks.

---

## 8. Disadvantages

### Dying ReLU Problem

If a neuron consistently receives negative inputs:

$$
x < 0
$$

then

$$
\text{ReLU}(x) = 0
$$

and

$$
\text{ReLU}'(x) = 0
$$

As a result, the neuron may stop learning permanently.

---

## 9. Variants of ReLU

### Leaky ReLU

Allows a small gradient for negative inputs:

$$
\text{Leaky ReLU}(x) =
\begin{cases}
0.01x, & x < 0 \
x, & x \geq 0
\end{cases}
$$

### Parametric ReLU (PReLU)

$$
\text{PReLU}(x) =
\begin{cases}
ax, & x < 0 \
x, & x \geq 0
\end{cases}
$$

where $a$ is learned during training.

### ELU (Exponential Linear Unit)

$$
\text{ELU}(x)=
\begin{cases}
x, & x \geq 0 \
\alpha(e^x-1), & x < 0
\end{cases}
$$

---

## 10. Comparison with Sigmoid and Tanh

| Feature            | Sigmoid | Tanh      | ReLU         |
| ------------------ | ------- | --------- | ------------ |
| Output Range       | $(0,1)$ | $(-1,1)$  | $[0,\infty)$ |
| Zero-Centered      | No      | Yes       | No           |
| Vanishing Gradient | Severe  | Moderate  | Minimal      |
| Computational Cost | High    | High      | Very Low     |
| Hidden Layer Usage | Rare    | Sometimes | Most Common  |

---

## Summary

The **ReLU (Rectified Linear Unit)** is defined as:

$$
\text{ReLU}(x) = \max(0, x)
$$

It is currently the most popular activation function for hidden layers because it is simple, fast, and helps mitigate the vanishing gradient problem. Its primary limitation is the **Dying ReLU Problem**, which can be addressed using variants such as **Leaky ReLU** and **PReLU**.

# 🔰Huber Loss Function🔰

The **Huber Loss** is a loss function used in regression problems. It combines the advantages of **Mean Squared Error (MSE)** and **Mean Absolute Error (MAE)**.

* For small errors, it behaves like **MSE**.
* For large errors (outliers), it behaves like **MAE**.

This makes Huber Loss more robust to outliers than MSE while remaining differentiable.

---

## 1. Mathematical Formula

Let:

$$
y
$$

be the true value and

$$
\hat{y}
$$

be the predicted value.

The prediction error is:

$$
e = y - \hat{y}
$$

The Huber Loss is defined as:

$$
L_{\delta}(e) =
\begin{cases}
\frac{1}{2}e^2, & |e| \leq \delta [8pt]
\delta\left(|e|-\frac{\delta}{2}\right), & |e| > \delta
\end{cases}
$$

where:

* $e$ is the prediction error.
* $\delta$ is a threshold parameter.

---

## 2. Interpretation

### Small Errors

When:

$$
|e| \leq \delta
$$

the loss becomes:

$$
L(e) = \frac{1}{2}e^2
$$

which is identical to MSE.

The curve is smooth and quadratic near zero.

### Large Errors

When:

$$
|e| > \delta
$$

the loss becomes:

$$
L(e)=\delta\left(|e|-\frac{\delta}{2}\right)
$$

which grows linearly like MAE.

This prevents extremely large errors from dominating the loss.

---

## 3. Example (δ = 1)

### Error = 0.5

Since:

$$
|0.5| \leq 1
$$

$$
L(0.5)=\frac{1}{2}(0.5)^2
$$

$$
L(0.5)=0.125
$$

### Error = 3

Since:

$$
|3| > 1
$$

$$
L(3)=1\left(3-\frac{1}{2}\right)
$$

$$
L(3)=2.5
$$

---

## 4. Derivative of Huber Loss

The gradient is:

$$
\frac{dL}{de}
$$
=============
$$
\begin{cases}
e, & |e| \leq \delta [8pt]
\delta , \text{sign}(e), & |e| > \delta
\end{cases}
$$

where

$$
\text{sign}(e)=
\begin{cases}
1, & e>0 \
0, & e=0 \
-1, & e<0
\end{cases}
$$

---

## 5. Advantages

* Less sensitive to outliers than MSE.
* Smooth and differentiable.
* Stable gradients during training.
* Combines strengths of MSE and MAE.

---

## 6. Disadvantages

* Requires choosing the threshold parameter $\delta$.
* Slightly more computationally expensive than MSE.
* Performance depends on selecting a suitable $\delta$.

---

## 7. Comparison with MSE and MAE

| Property               | MSE        | MAE               | Huber Loss         |
| ---------------------- | ---------- | ----------------- | ------------------ |
| Formula Growth         | Quadratic  | Linear            | Quadratic + Linear |
| Sensitive to Outliers  | High       | Low               | Moderate           |
| Differentiable at Zero | Yes        | No                | Yes                |
| Stable Gradients       | No         | Yes               | Yes                |
| Common Use             | Regression | Robust Regression | Robust Regression  |

---

## 8. When to Use Huber Loss

Use Huber Loss when:

* The dataset contains outliers.
* You want robustness without completely ignoring large errors.
* MSE is overly affected by extreme values.
* MAE produces unstable optimization.

It is widely used in:

* Regression tasks
* Time-series forecasting
* Reinforcement learning
* Object detection models

---

## Summary

The **Huber Loss** combines the benefits of MSE and MAE:

$$
L_{\delta}(e)=
\begin{cases}
\frac{1}{2}e^2, & |e| \leq \delta [8pt]
\delta\left(|e|-\frac{\delta}{2}\right), & |e| > \delta
\end{cases}
$$

* **Quadratic** for small errors.
* **Linear** for large errors.
* **Robust to outliers**.
* **Smooth and differentiable**.

Because of its balance between sensitivity and robustness, Huber Loss is often preferred for regression problems with noisy data.

# Momentum in Gradient Descent

## What is Momentum?

Momentum is an optimization technique used to speed up Gradient Descent and reduce oscillations during training.

Instead of moving only according to the current gradient, Momentum also considers the previous update direction.

---

## Why Momentum?

Without Momentum:

- Slow convergence
- Zig-zag movement
- Can get stuck in valleys

With Momentum:

- Faster convergence
- Smoother path to the minimum
- Reduced oscillations

---

## Intuition

Imagine a ball rolling down a hill.

- Gradient Descent: The ball checks the slope at every step.
- Momentum: The ball gains speed as it rolls and keeps moving in the same direction.

This helps reach the minimum faster.

---

## Momentum Update Equations

Velocity update:

$$
v_t = \beta v_{t-1} - \eta \nabla J(\theta_t)
$$

Parameter update:

$$
\theta_{t+1} = \theta_t + v_t
$$

Where:

- $v_t$ = current velocity
- $v_{t-1}$ = previous velocity
- $\beta$ = momentum coefficient
- $\eta$ = learning rate
- $\nabla J(\theta_t)$ = gradient of loss function
- $\theta$ = model parameters

---

## Typical Value of Momentum

$$
\beta = 0.9
$$

Common values:

- 0.8
- 0.9
- 0.99

---

## Comparison

| Gradient Descent | Momentum |
|-----------------|-----------|
| Uses current gradient only | Uses current + past gradients |
| Slower | Faster |
| More oscillation | Less oscillation |
| May converge slowly | Faster convergence |

---

## Example

Suppose:

$$
\eta = 0.1
$$

$$
\beta = 0.9
$$

Previous velocity:

$$
v_{t-1} = 0.5
$$

Current gradient:

$$
\nabla J(\theta_t) = 2
$$

Velocity:

$$
v_t = 0.9(0.5) - 0.1(2)
$$

$$
v_t = 0.45 - 0.2
$$

$$
v_t = 0.25
$$

Parameter update:

$$
\theta_{t+1} = \theta_t + 0.25
$$

---

## Advantages

1. Faster convergence
2. Reduced oscillations
3. Better optimization
4. Helps in deep neural networks

# AdaGrad (Adaptive Gradient)

## What is AdaGrad?

AdaGrad (Adaptive Gradient) is an optimization algorithm introduced in 2011.

It was the first widely-used adaptive learning rate optimizer and is the predecessor of:

- RMSProp
- Adam

AdaGrad automatically adjusts the learning rate for each parameter.

---

# Why AdaGrad?

In standard Gradient Descent:

$$
\theta = \theta - \eta g
$$

The learning rate $\eta$ remains constant.

In AdaGrad:

- Frequently updated parameters get smaller learning rates.
- Rarely updated parameters get larger learning rates.

This makes AdaGrad particularly useful for:

- NLP
- Sparse data
- Recommendation systems

---

# Stalling Problem

Notice:

$$
G_t
=
\sum_{i=1}^{t} g_i^2
$$

Since every squared gradient is added forever:

$$
G_t \uparrow
$$

Therefore:

$$
\sqrt{G_t}\uparrow
$$

And:

$$
\frac{\eta}{\sqrt{G_t}}
\downarrow
$$

Eventually:

$$
\eta_{eff}
\approx 0
$$

Training almost stops.

This is called the **AdaGrad Stalling Problem**.

---

# Advantages

1. Adaptive learning rates
2. No manual tuning for each parameter
3. Excellent for sparse data
4. Useful in NLP applications

---

# Disadvantages

1. Learning rate keeps shrinking
2. Training eventually slows down
3. Poor performance on dense datasets
4. Suffers from stalling

---

# How RMSProp Fixes AdaGrad

AdaGrad:

$$
G_t = G_{t-1} + g_t^2
$$

Stores all gradients forever.

RMSProp:

$$
G_t = \beta G_{t-1} + (1-\beta)g_t^2
$$

Uses an Exponential Moving Average (EMA).

This prevents unlimited growth of $G_t$.

# RMSProp (Root Mean Square Propagation)

## What is RMSProp?

RMSProp is an adaptive optimization algorithm developed to solve the major drawback of AdaGrad.

AdaGrad keeps accumulating all past squared gradients:

$$
G_t = G_{t-1} + g_t^2
$$

As a result:

$$
G_t \uparrow
$$

and

$$
\frac{\eta}{\sqrt{G_t}}
\downarrow
$$

The learning rate eventually becomes very small, causing training to stall.

RMSProp solves this problem by using an Exponential Moving Average (EMA) of squared gradients.

# 🔰RMSProp (Root Mean Square Propagation)🔰

## Motivation

AdaGrad accumulates all past squared gradients:

$$
G_t = G_{t-1} + g_t^2
$$

As training progresses:

$$
G_t \uparrow
$$

Therefore:

$$
\frac{\eta}{\sqrt{G_t}}
\downarrow
$$

The learning rate becomes extremely small and training eventually stalls.

RMSProp was introduced to solve this problem.

---

# Key Idea

Instead of storing all past squared gradients forever, RMSProp uses an Exponential Moving Average (EMA).

Old gradients gradually fade away.

This prevents the denominator from growing without bound.

- $\eta$ = learning rate
