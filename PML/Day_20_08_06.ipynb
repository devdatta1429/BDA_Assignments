```python
df = pd.read_excel('CDAC_DataBook.xlsx', sheet_name = 'diabetes')

df =df[['Glucose','BloodPressure','Age','DietType','Outcome']]

x = df.drop('Outcome', axis=1)
y = df.Outcome

diet_dummy = pd.get_dummies(x.DietType, drop_first=True, prefix='Diet').astype(int)
x = x.drop('DietType', axis=1)
x = pd.concat([x, diet_dummy], axis=1)

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=20)

x_train = sm.add_constant(x_train, prepend=False)
mod3 = sm.Logit(y_train, x_train).fit()
print(mod3.summary())
```

```python
y_pred = mod3.predict(x_train)
y_probs = mod3.predict(x_train)
y_pred = []
for ctr in y_probs:
    if ctr >= 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)

print(confusion_matrix(y_train, y_pred))
```

# Confusion Matrix Metrics Notes

## Confusion Matrix

\[
\begin{bmatrix}
TN & FP \\
FN & TP
\end{bmatrix}
=
\begin{bmatrix}
362 & 37 \\
51 & 164
\end{bmatrix}
\]

Where:

- **TN (True Negative) = 362**
  - Model correctly predicted Negative class.

- **FP (False Positive) = 37**
  - Model predicted Positive, but actual class was Negative.

- **FN (False Negative) = 51**
  - Model predicted Negative, but actual class was Positive.

- **TP (True Positive) = 164**
  - Model correctly predicted Positive class.

---

# 1. Precision (Positive Class)

### Definition
Precision measures how many of the predicted positive cases are actually positive.

### Formula

\[
Precision = \frac{TP}{TP + FP}
\]

### Calculation

\[
Precision = \frac{164}{164 + 37}
\]

\[
Precision = \frac{164}{201}
\]

\[
Precision = 0.8159
\]

### Result

**Precision = 81.59%**

---

# 2. Specificity (Negative Class Precision)

### Definition
Measures how accurately the model identifies negative cases.

### Formula

\[
Specificity = \frac{TN}{TN + FP}
\]

### Calculation

\[
Specificity = \frac{362}{362 + 37}
\]

\[
Specificity = \frac{362}{399}
\]

\[
Specificity = 0.9073
\]

### Result

**Specificity = 90.73%**

---

# 3. Recall (Sensitivity / True Positive Rate)

### Definition
Recall measures how many actual positive cases are correctly identified by the model.

### Formula

\[
Recall = \frac{TP}{TP + FN}
\]

### Calculation

\[
Recall = \frac{164}{164 + 51}
\]

\[
Recall = \frac{164}{215}
\]

\[
Recall = 0.7628
\]

### Result

**Recall = 76.28%**

---

# 4. Negative Predictive Value (NPV)

### Definition
Measures how many predicted negative cases are actually negative.

### Formula

\[
NPV = \frac{TN}{TN + FN}
\]

### Calculation

\[
NPV = \frac{362}{362 + 51}
\]

\[
NPV = \frac{362}{413}
\]

\[
NPV = 0.8808
\]

### Result

**NPV = 88.08%**

---

# Summary Table

| Metric | Formula | Value |
|----------|----------|----------|
| Precision | TP / (TP + FP) | 0.8159 (81.59%) |
| Recall (Sensitivity) | TP / (TP + FN) | 0.7628 (76.28%) |
| Specificity | TN / (TN + FP) | 0.9073 (90.73%) |
| Negative Predictive Value (NPV) | TN / (TN + FN) | 0.8808 (88.08%) |

---

# Interpretation

- **Precision (81.59%)**: Out of all samples predicted as positive, 81.59% were actually positive.
- **Recall (76.28%)**: The model correctly detected 76.28% of all actual positive samples.
- **Specificity (90.73%)**: The model correctly identified 90.73% of actual negative samples.
- **NPV (88.08%)**: Out of all samples predicted as negative, 88.08% were actually negative.

The model performs better at identifying **negative cases (Specificity = 90.73%)** than **positive cases (Recall = 76.28%)**.
