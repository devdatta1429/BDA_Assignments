
# 🔰KNN🔰


## K-Nearest Neighbors (KNN)

- KNN (K-Nearest Neighbors) is a **distance-based supervised learning algorithm** used for classification and regression.
- It predicts the output based on the **K nearest data points** to the query point.
- Since KNN relies on distance calculations, **feature scaling is necessary** to ensure that features with larger values do not dominate the distance measure.
- Common scaling techniques include **Standardization** and **Min-Max Normalization**.
- The choice of \(k\) affects performance:
  - Small \(k\): sensitive to noise and overfitting.
  - Large \(k\): may lead to underfitting.
- Common distance metric:

\
$$
d(x,y)=\sqrt{\sum_{i=1}^{n}(x_i-y_i)^2}
$$

where \(d(x,y)\) is the Euclidean distance between two points.

### Advantages
- Simple and easy to implement.
- No training phase required.
- Works well for small datasets.

### Limitations
- Computationally expensive for large datasets.
- Sensitive to irrelevant features and outliers.
- Requires proper feature scaling.

# Support Vector Machine (SVM)

## What is SVM?

Support Vector Machine (SVM) is a Supervised Machine Learning algorithm used for:

1. Classification
2. Regression

Its main objective is to find the best boundary (hyperplane) that separates different classes.

---

# Intuition

Suppose we have two classes:

```text
● ● ● ● ●

○ ○ ○ ○ ○
```

Many lines can separate them.

SVM chooses the line that leaves the maximum distance between the two classes.

This distance is called the Margin.

---

# Hyperplane

A Hyperplane is the decision boundary that separates classes.

### In 2D

Hyperplane = Line

```text
● ● ● ●

-----------
○ ○ ○ ○
```

### In 3D

Hyperplane = Plane

### In Higher Dimensions

Hyperplane = N-Dimensional Surface

---

# Margin

Margin is the distance between the hyperplane and the nearest data points.

SVM tries to maximize this margin.

### Why?

Larger margin:

- Better generalization
- Lower overfitting
- Better classification

---

# Support Vectors

Support Vectors are the nearest data points to the hyperplane.

They are the most important points in the dataset.

Removing other points may not affect the decision boundary.

Removing support vectors changes the hyperplane.

---

# SVM Goal

Find the hyperplane with:

$$
\text{Maximum Margin}
$$

between classes.

---

# Equation of Hyperplane



Where:

- $w$ = weight vector
- $x$ = input features
- $b$ = bias

---

# Classification Rule

If:

$$
w^Tx+b > 0
$$

Class = Positive

If:

$$
w^Tx+b < 0
$$

Class = Negative

---

# Margin Equations

Positive Class Boundary:

$$
w^Tx+b=1
$$

Negative Class Boundary:

$$
w^Tx+b=-1
$$

Distance between them:

$$
\text{Margin}
=
\frac{2}{||w||}
$$

---

# Types of SVM

## 1. Linear SVM

Used when data is linearly separable.

Example:

```text
● ● ● ●

-----------
○ ○ ○ ○
```

A straight line can separate classes.

---

## 2. Non-Linear SVM

Used when data cannot be separated by a straight line.

Example:

```text
○ ○ ○ ○

● ● ●

○ ○ ○ ○
```

Need a curved boundary.

---

# Kernel Trick

Kernel Trick transforms data into a higher-dimensional space where it becomes linearly separable.

Instead of explicitly transforming data, kernels compute similarity directly.

---

# Common Kernels

## 1. Linear Kernel

$$
K(x,y)=x^Ty
$$

Used for linearly separable data.

---

## 2. Polynomial Kernel

$$
K(x,y)=(x^Ty+c)^d
$$

Captures polynomial relationships.

---

## 3. RBF (Gaussian) Kernel

Most popular kernel.

$$
K(x,y)=e^{-\gamma ||x-y||^2}
$$

Used for complex non-linear boundaries.

---

## 4. Sigmoid Kernel

$$
K(x,y)=\tanh(ax^Ty+b)
$$

Inspired by neural networks.

---

# Parameter C (Regularization)

C controls the tradeoff between:

- Margin size
- Classification error

---

## Large C

```text
C = 100
```

- Smaller margin
- Fewer misclassifications
- May overfit

---

## Small C

```text
C = 0.1
```

- Larger margin
- More tolerance for errors
- Better generalization

---

# SVM Epsilon (ε) — Used in SVR (Support Vector Regression)

Many students get confused because:

- **SVM (Classification)** mainly uses **C**
- **SVR (Support Vector Regression)** uses **C** and **ε (epsilon)**

---

## What is Epsilon (ε)?

Epsilon defines a **tolerance zone** around the regression line.

This zone is called the **Epsilon Tube**.

Inside this tube:

- ✅ Errors are ignored

Outside this tube:

- ❌ Errors are penalized

---

## Epsilon Tube

```text
         Upper Boundary
         y = f(x) + ε
         ----------------

               ●

Regression Line
y = f(x)
----------------

        ●

         ----------------
         y = f(x) - ε

         Lower Boundary
```

Points inside the tube:

```text
No Loss
```

Points outside the tube:

```text
Loss is Calculated
```

---

## SVR Objective

Instead of minimizing every error, SVR says:

> "I don't care about small errors within ε."

This makes the model more robust to noise.

---

## Example 1

Actual Value:

$$
y = 100
$$

Predicted Value:

$$
\hat{y} = 102
$$

Error:

$$
|100 - 102| = 2
$$

If:

$$
\epsilon = 5
$$

Since:

$$
2 < 5
$$

**No penalty is applied.**

---

## Example 2

Actual Value:

$$
y = 100
$$

Predicted Value:

$$
\hat{y} = 110
$$

Error:

$$
|100 - 110| = 10
$$

If:

$$
\epsilon = 5
$$

Then:

$$
10 - 5 = 5
$$

**Penalty is applied.**

---

## Visualization

```text
              |
              |     *
              |
 y=f(x)+ε ----|-------------------
              |
              |    *
              |
 y=f(x) ------|-------------------
              |
              |  *
              |
 y=f(x)-ε ----|-------------------
              |
```

All points between:

$$
f(x)-\epsilon
$$

and

$$
f(x)+\epsilon
$$

are accepted without penalty.

---

## Effect of Epsilon

### Small ε

```text
ε = 0.1
```

- Narrow tube
- More points outside
- More support vectors
- Complex model
- Risk of overfitting

### Large ε

```text
ε = 10
```

- Wide tube
- Fewer support vectors
- Simpler model
- Risk of underfitting

---

## Relationship Between C and ε

| Parameter | Purpose |
|-----------|---------|
| **C** | Error Penalty |
| **ε** | Error Tolerance |

### C
Controls how much the model penalizes errors.

### ε
Controls how much error is acceptable without penalty.

---

## Interview Question

### What is Epsilon in SVR?

> Epsilon (ε) defines a margin of tolerance around the regression function. Errors within this epsilon tube are ignored, while errors outside the tube are penalized. This helps SVR become robust to small variations and noise in the data.

---

## Memory Trick

- **C = How much you punish errors**
- **ε = How much error you are willing to tolerate**

For SVR:

```text
Small ε  → Strict Teacher
Large ε  → Lenient Teacher
```

A strict teacher marks every small mistake, while a lenient teacher ignores small mistakes. This is exactly how ε works in Support Vector Regression.
