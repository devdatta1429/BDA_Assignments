
# 🔰Principal Component Analysis (PCA)🔰

## PCA
- PCA = **Dimensionality Reduction** technique.
- Converts many features into fewer **Principal Components**.
- Retains maximum information (variance).

## Curse of Dimensionality
- As the number of features increases:
  - Data becomes sparse.
  - Computation increases.
  - Finding meaningful patterns becomes difficult.
- This problem is called the **Curse of Dimensionality**.

## Pairwise Feature Combinations

Formula:

$$
\frac{n(n-1)}{2}
$$

where \(n\) = number of variables.

### Examples

For \(n=7\):

$$
\frac{7(7-1)}{2}=21
$$

➡️ 7 variables → 21 combinations

For \(n=10\):

$$
\frac{10(10-1)}{2}=45
$$

➡️ 10 variables → 45 combinations

## PCA Goal
- Reduce dimensions.
- Preserve maximum variance.
- Remove redundancy.
- Improve model efficiency.

## Key Point
More features ⇒ More combinations ⇒ More complexity ⇒ PCA helps reduce this complexity.



```python
df = pd.read_excel('pca.xlsx')

x = df.drop('Response', axis=1)
y = df.Response

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=20)

x_train = sm.add_constant(x_train, prepend=False)

mod1 = sm.Logit(y_train,x_train).fit()

x_test = sm.add_constant(x_test, prepend=False)
y_pred = mod1.predict(x_test)
y_pred = (y_pred>=0.5).astype(int)

print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.decomposition import PCA
pca = PCA()

pca.fit(x_train)
pca.components_
```

```python
# .T means Transform
pca.components_

loadings = pd.DataFrame( pca.components_.T,
    columns=[f"PC{i+1}" for i in range(len(pca.components_))],
    index=x_train.columns
)

print(loadings)
```
