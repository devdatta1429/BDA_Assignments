- it tells you about the +ve or -ve corelation between two factors but nothe magnitude.
- Formula:
$$
Co-Variance (COV) = \frac{\Sigma^{N}_{i=1} (X_i - \mu_x)(Y_i - \mu_y)}{n - 1}
$$
$$
Co-Relation Coeffient(CORR) = \frac{COV}{\sigma_1 * \sigma_2}
$$

# Principal Component Analysis (PCA) - Manual Calculation Using Characteristic Equation

## Step 1: Given Dataset

| Feature 1 | Feature 2 | Feature 3 |
|------------|------------|------------|
| 2 | 6 | 8 |
| 4 | 4 | 5 |
| 5 | 3 | 9 |
| 2 | 3 | 9 |
| 7 | 8 | 7 |

Let

$$
X=
\begin{bmatrix}
2 & 6 & 8\\
4 & 4 & 5\\
5 & 3 & 9\\
2 & 3 & 9\\
7 & 8 & 7
\end{bmatrix}
$$

Number of observations:

$$
n=5
$$

Number of features:

$$
p=3
$$

---

# Step 2: Compute Mean of Each Feature

For Feature 1:

$$
\bar{x}_1=\frac{2+4+5+2+7}{5}
=\frac{20}{5}
=4
$$

For Feature 2:

$$
\bar{x}_2=\frac{6+4+3+3+8}{5}
=\frac{24}{5}
=4.8

For Feature 3:

$$

Mean Vector:

$$

\begin{bmatrix}
4\\
4.8\\
7.6

# Step 3: Mean Center the Dataset

Subtract column means from each feature.

$$
X_{centered}
=

$$

# Step 4: Compute Covariance Matrix

Formula:

$$
C=
\frac{1}{n-1}
X_{centered}^{T}
X_{centered}
$$

Since:

$$
n=5

$$
n-1=4
$$

Computing:

$$

\end{bmatrix}
$$

# Step 5: Find Eigenvalues Using Characteristic Equation

PCA requires solving:

$$
|C-\lambda I|=0
$$

Substitute covariance matrix:

$$
\left|
\begin{matrix}

\right|

$$

$$

$$

$$
-

\right]
$$

$$
+
(-0.5)

$$

0

```python
# Dataset Loading
df = pd.read_excel('pca.xlsx')
df.head()

# Train-Test Split
x = df.drop(columns=["Response"])
y = df.Response
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=20)

# Standardization
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# PCA model Fit
pca = PCA(n_components=2)
x_train_pc = pca.fit_transform(x_train_scaled)
x_test_pc = pca.transform(x_test_scaled)

# Creating DataFrame of max variation explained
x_train_pc = pd.DataFrame(x_train_pc, columns=['PC1','PC2'], index = x_train.index)
x_test_pc = pd.DataFrame(x_test_pc, columns=['PC1','PC2'], index = x_test.index)
```

```python
# Model on PC vectors
x_train_pc1 = sm.add_constant(x_train_pc, prepend=False)
x_test_pc1 = sm.add_constant(x_test_pc, prepend=False)

mod2pca = sm.Logit(y_train,x_train_pc1).fit()

y_pred = mod2pca.predict(x_test_pc1)
y_pred = (y_pred>=0.5).astype(int)
print(confusion_matrix(y_test,y_pred))
```
