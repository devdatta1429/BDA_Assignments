# 📊 Covariance & Correlation Notes (Easy Version)

---

### 📊 Given Data

| Index | X values | Y values |
|------|----------|----------|
| 1 | 5 | 3 |
| 2 | 4 | 12 |
| 3 | 7 | 9 |
| 4 | 6 | 7 |
| 5 | 8 | 6 |
| 6 | 9 | 8 |


---

## 📌 Step 1: Covariance

Covariance tells us:
> How two variables move together

### Result:
Cov(X, Y) = **-0.9**

### Interpretation:
- Negative covariance → inverse relationship
- When X increases, Y tends to decrease
- So this shows a **negative relationship**

---

## 📌 Step 2: Standard Deviation

We calculate:

- Std(X) = standard deviation of [5, 4, 7, 6, 8, 9]
- Std(Y) = standard deviation of [3, 12, 9, 7, 6, 8]

These measure spread of each dataset.

---

## 📌 Step 3: Correlation

### Formula:
$$
corr(X,Y) = \frac{cov(X,Y)}{\sigma_X \cdot \sigma_Y}
$$
---

### Substituting values:

$$
corr = \frac{-0.9}{(std(5,4,7,6,8,9) \times std(3,12,9,7,6,8))}
$$

---

### Final Result:

corr = **0.15947**

---

## 📌 Interpretation of Correlation

| Value | Meaning |
|------|--------|
| +1 | Strong positive relationship |
| 0 | No relationship |
| -1 | Strong negative relationship |

---

### Our Result:
- corr = **0.15947**
- Very close to 0
- So this is a **very weak relationship**

---

## ⚠️ Important Insight

Even though:
- Covariance = **-0.9** (negative direction)

After normalization:
- Correlation becomes **very small**

👉 This means:
- Relationship is **weak**
- Not strongly linear

---

## 📌 Final Summary

- Covariance = **-0.9** → negative trend
- Correlation = **0.15947** → very weak relationship
- Overall: **weak / almost no linear relationship**
