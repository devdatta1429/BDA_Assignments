# 🔰 Convolutional Neural Network (CNN)

## What is CNN?

* **CNN (Convolutional Neural Network)** is an extension of **ANN (Artificial Neural Network)**.
* It is mainly used for **image processing and computer vision tasks**.
* Images have a **spatial structure**, meaning nearby pixels are related to each other.
* ANN destroys this relationship by flattening the image into a 1D vector.
* CNN preserves spatial relationships and learns important patterns such as edges, textures, and shapes.

---

# CNN Pipeline

```text
Image
  ↓
Convolution
  ↓
ReLU
  ↓
Pooling
  ↓
(Repeat Convolution + ReLU + Pooling Blocks)
  ↓
Flatten
  ↓
Dense Layer
  ↓
Output
```

---

# 1️⃣ Convolution Layer

The convolution layer applies a **filter (kernel)** to the image to extract important features.

### Example

| Image Size | Filter Size | Output Size |
| ---------- | ----------- | ----------- |
| 6 × 6      | 3 × 3       | 4 × 4       |

### Output Size Formula (Without Padding)

$$
\text{Output Size} = (N - F) + 1
$$

Where:

* $N$ = Size of input image
* $F$ = Size of filter

### Example

$$
(6 - 3) + 1 = 4
$$

Therefore,

$$
6 \times 6 \rightarrow 4 \times 4
$$

---

# 2️⃣ ReLU (Rectified Linear Unit)

ReLU introduces non-linearity into the network.

### Formula

$$
f(x)=\max(0,x)
$$

### Working

```text
x > 0  → x
x ≤ 0  → 0
```

### Advantages

* Faster training
* Reduces vanishing gradient problem
* Helps learn complex patterns

---

# 3️⃣ Pooling Layer

Pooling reduces the size of feature maps while keeping important information.

## Types of Pooling

### Max Pooling

Selects the maximum value from a region.

Example:

```text
2 5
1 3

Output = 5
```

### Average Pooling

Computes the average value of a region.

### Benefits

* Reduces computation
* Reduces overfitting
* Retains important features

---

# 4️⃣ Flatten Layer

Converts a 2D feature map into a 1D vector.

Example:

```text
1 2
3 4

↓

[1, 2, 3, 4]
```

---

# 5️⃣ Dense Layer

* Fully Connected Layer
* Similar to ANN layers
* Combines extracted features for prediction

---

# 6️⃣ Output Layer

Produces the final prediction or classification result.

Examples:

* Cat vs Dog Classification
* Digit Recognition
* Object Detection

---

# Padding

Padding adds extra pixels (usually zeros) around the image border.

### Why Padding?

* Prevents loss of edge information
* Controls output size
* Helps preserve spatial dimensions

---

## Output Size Formula (With Padding)

$$
\text{Output Size}
==================

\frac{N - F + 2P}{S} + 1
$$

Where:

* $N$ = Input size
* $F$ = Filter size
* $P$ = Padding
* $S$ = Stride

### Example

Given:

$$
N=6,\quad F=3,\quad P=1,\quad S=1
$$

$$
\text{Output Size}
==================

\frac{6-3+2(1)}{1}+1
=6
$$

Thus,

```text
Input  : 6 × 6
Filter : 3 × 3
Padding: 1
Output : 6 × 6
```

# Max Pooling Layer

## What is Max Pooling?

Max Pooling is a down-sampling operation used in CNNs to reduce the spatial dimensions of feature maps while retaining the most important features.

Instead of keeping every value, Max Pooling selects the **maximum value** from a small region (window) of the feature map.

### Why Max Pooling?

- Reduces the size of feature maps
- Reduces computational cost
- Reduces overfitting
- Retains the most important features
- Makes the model more robust to small shifts in the image

---

## Example: Max Pooling (2×2, Stride = 2)

### Input Feature Map (6×6)

$$
\begin{bmatrix}
1&3&2&4&5&1\\
5&6&1&2&7&3\\
7&2&8&1&4&6\\
4&5&3&9&2&8\\
1&7&4&2&6&3\\
8&2&5&1&9&4
\end{bmatrix}
$$

### Max Pooling Operation

A **2×2 window** moves across the feature map with **stride = 2**.

For each window, the maximum value is selected.

```text id="ftxqwt"
Window 1          Window 2          Window 3

1 3               2 4               5 1
5 6               1 2               7 3

Max=6             Max=4             Max=7
```

```text id="0j0m3u"
Window 4          Window 5          Window 6

7 2               8 1               4 6
4 5               3 9               2 8

Max=7             Max=9             Max=8
```

```text id="wwrnz7"
Window 7          Window 8          Window 9

1 7               4 2               6 3
8 2               5 1               9 4

Max=8             Max=5             Max=9
```

### Output Feature Map (3×3)

$$
\begin{bmatrix}
6&4&7\\
7&9&8\\
8&5&9
\end{bmatrix}
$$

---

## Complete Transformation

$$
\begin{bmatrix}
1&3&2&4&5&1\\
5&6&1&2&7&3\\
7&2&8&1&4&6\\
4&5&3&9&2&8\\
1&7&4&2&6&3\\
8&2&5&1&9&4
\end{bmatrix}
\qquad
\xrightarrow{\text{Max Pooling (2×2)}}
\qquad
\begin{bmatrix}
6&4&7\\
7&9&8\\
8&5&9
\end{bmatrix}
$$

---

## What is Happening Here?

The pooling layer scans the feature map using a 2×2 window:

### First Window

$$
\begin{bmatrix}
1&3\\
5&6
\end{bmatrix}
$$

Maximum value:

$$
\max(1,3,5,6)=6
$$

Output:

$$
6
$$

---

### Second Window

$$
\begin{bmatrix}
2&4\\
1&2
\end{bmatrix}
$$

Maximum value:

$$
4
$$

Output:

$$
4
$$

The same process continues for all windows.

---

## Output Size Formula

$$
\text{Output Size}
=
\frac{N-F}{S}+1
$$

Where:

- $N$ = Input Size
- $F$ = Pooling Window Size
- $S$ = Stride

For this example:

$$
N=6,\quad F=2,\quad S=2
$$

$$
\text{Output Size}
=
\frac{6-2}{2}+1
=
3
$$

Therefore:

$$
6\times6 \rightarrow 3\times3
$$

---

## Key Idea

Max Pooling keeps only the **strongest activation** from each region.

```text id="ob5p20"
Before Pooling : 36 values (6×6)

After Pooling  : 9 values (3×3)
```

Thus, the network focuses on the most important features while reducing memory and computation.
