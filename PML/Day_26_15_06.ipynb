Q1. What does Activation Function do?
- Activation Function is used for introducing the non-linearity.

Q2. What Accumulation Function do?
- Accumulation Function is summation of all other input which are provided to nueron.

Forward Propagation:
- We we move from the one input to ouput layer through hidden layer
- when the output is compared with the true value we determine the losses, to reduce the losses we perform the back propagation.

Back Propagation:
- During Back propagation weight of each nueron will get updated.

Weight updation Formula:
$$
w_{t+1(new)} = w_{t(old)} - \eta \frac{\partial L}{\partial w_t}
$$

Where: \
$w_t$= current weight \
$w_{t+1}$= updated weight \
η = learning rate \
L = loss function \
$\frac{∂w}{∂L}$ = gradient of the loss with respect to the weight
