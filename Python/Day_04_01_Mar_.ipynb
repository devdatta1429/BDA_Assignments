```python
#Shallow Copy

data=[1,2,3]
d1=data

print(data)  # [1, 2, 3]
print(d1)    # [1, 2, 3]

d1[0]=5
print(data)  # [5, 2, 3]
print(d1)    # [5, 2, 3]

print(id(data)) # 132352657582272
print(id(d1))   # 132352657582272
```

```python
#Deep Copy

data=[1,2,3]
d1=data.copy()
d1[0]=5
print(data)  # [1, 2, 3]
print(d1)    # [1, 2, 3]

print(id(data)) # 132352657583424
print(id(d1))   # 132352657583232
```

```python
def smaller(num=1):
  data=[10,20,30,25,35,45,65,50,60,80,70]
  return sorted(data)[num-1]

smaller(0)

def larger(num):
  data=[10,20,30,25,35,45,65,50,60,80,70]
  return sorted(data)[-(num)]

larger(1)
```

```python
shop_list=[]

def add(num):
  return shop_list.append(num)

def dele (num):

  if num in shop_list:
        shop_list.remove(num)
        print(f"Removed {num} from the list.")
  else:
        print(f"{num} is not in your list.")

def all():
    print(shop_list)

def all_clear():
    shop_list.clear()
    print(shop_list)
    print("All items deleted")



while True:
  comm=int(input('''
  ----------------------
  Enter your Comm :
  1. add
  2. delete
  3. all value
  4. delete all
  5. Exit
  -----------------------
  '''))


  if comm==1:
    value=(input("Enter the Value"))
    add(value)
    for i in shop_list:
        print(f"{(shop_list.index(i))+1} {i}")
  elif comm==2:
    if len(shop_list) == 0:
      print("The list is empty!")
    else:
      value=(int(input("Enter the Value")))
      dele((value.index())-1)
    print(shop_list)
  elif comm==3:
    all()
  elif comm==4:
    if len(shop_list) == 0:
        print("The list is empty!")
    else:
        all_clear()
  elif comm==5:
    break
  else:
    print("Enter wrong Output")

print(f"This is full list{shop_list}")


```

```python
data=[10,20,30,40,50]

def rotate(n):
    data.append (data.pop(0))
    if n==0:
        return 1
    else:
        return n * rotate(n-1)

rotate(4)
print(data)
```
