```python
class Time:
    def set_time(self,HH,MM,SS):
        self.HH=HH
        self.MM=MM
        self.SS=SS

    def display(self):
        print(self.HH,self.MM,self.SS)

    def in_HH(self):
        self.HH+=1

    def in_MM(self):
        self.MM+=1

    def in_SS(self):
        self.SS+=1
```

```python
t = Time()
t.set_time(10, 20, 30)

t.display()

t.in_HH()
t.in_MM()
t.in_SS()

t.display()
```

```python



class education:
    def __init__(self,education):
        self.education=education
        print(f"{education}")

class salary(education):
    def __init__(self,education,salary):
        self.salary = salary
        super().__init__(education)
        print(f"{salary}")

class human(salary):
    def __init__(self,name,gender,education,salary):
        self.name=name
        self.gender=gender
        super().__init__(education,salary)

    def display(self):
        print(f"{self.name}{self.gender}")
        print(f"{self.education}")
        print(f"{self.salary}")

h=human("dev","male","be",100000)
h.display()
```
