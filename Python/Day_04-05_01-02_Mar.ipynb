```python
#Deep Copy

data=[1,2,3]
d1=data.copy()
d1[0]=5
print(data)  # [1, 2, 3]
print(d1)    # [1, 2, 3]

print(id(data)) # 132352657583424
print(id(d1))   # 132352657583232
```
