```python
def rental(day, cost, ins=100):
    if day > 7:
        return ((cost * day) + (day * ins)) * 0.95
    else:
        return ((cost * day) + (day * ins))

day = int(input("Enter num of days: "))
cost = int(input("Enter rental cost: "))
ins_input = input("Enter Insurance (press Enter for default 100): ")

# Check if the user actually typed something
if ins_input == "":
    # We call the function WITHOUT the third argument to use the default
    total = rental(day, cost)
else:
    # We convert the input and pass it manually
    total = rental(day, cost, int(ins_input))

print(f"Total day: {day}, Daily Cost: {cost}, Total payable: {total}")
```

<H4>1️⃣ ATM Banking System</H4>

```python
balance=0

def Check_balance():
  return print(f"Balance : {balance}")

def deposit(num):

  global balance
  balance+=num
  return balance

def withdraw(num):
  global balance
  balance-=num

def exit():
  pass
  # break
print("----Welcome, To ATM----")
while True:
  print("1. Check Balance")
  print("2. Deposit Money")
  print("3. Withdraw Money")
  print("4. Exit")

  print("-------------------------")
  comm=int(input("Enter Command :"))
  if comm==1:
    print(Check_balance())
  elif comm==2:
    depo=int(input("enter your deposit Amount"))
    print(f"Your updated balance :{deposit(depo)}")
  elif comm==3:
    wd=int(input("enter your withdraw Amount"))
    if wd>balance:
      print("Your balance is low!!")
    else:
      print(f"updated Balance :{withdraw(wd)}")
  elif comm==4:
    break
  else:
    print("giving Wrong Input")
```

<H4>2️⃣ Student Result Management</H4>

```python
total=0

def percentage(marks):
  return marks/5

def grade(avg):
  if avg>75:
     return "Distinction"
  if avg>=60:
     return "First Class"
  if avg>=50:
    return "Second class"
  if avg>=40:
    return "Pass"
  else:
    return "fail"


print("-----Result Calculator----")
mark=map(int,input("Enter the value :").split())
mark1=list(mark)
for i in mark1:
  total+=i

  # print(total)

avg23= percentage(total)
avg=int(avg23)
print(f"Total Marks :{total}")
print(f"Percentage :{percentage(total)}")
print(f"Grade :{grade(avg)}")

# 78 67 89 56 90
```

```python
total=0

def bill(total):
  if total>5000:
    return total * 0.9
  return total

print()
while True:
  comm=int(input('''
  ----------------------
  Enter your Comm :
  1. Laptop : 45000
  2. mouse  : 1000
  3. Keyboard 2000
  4. Exit
  -----------------------
  '''))

  if comm==1:
    print("Laptop was added")
    total+=45000
  elif comm==2:
    print("Mouse was added")
    total+=1000
  elif comm==3:
    print("Keyboard was added")
    total+=2000
  elif comm==4:
    break
  else:
    print("Wrong Input")
  print(f"Cart Amount :{total}")

print(f"Total Price :{total}")
# print(f"Total Price :{total}")
print(f"Total Price :{bill(total)}")

```
