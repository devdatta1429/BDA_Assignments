```python
total_dist=0
ride_count=0
fare=50
while True:
  #print("-------------------------------")
  print("1. add a ride")
  print("2. 0 to Stop")
  print("-------------------------------")
  choice=int(input("Enter the choice :"))



  if choice==1:
    ride_count+=1
    dist=(int(input("Enter the distance :")))

    if dist<=20:
      current_fare=50+(dist*12)
    else:
      current_fare=50 +(dist*10.8)

    fare+=current_fare
    print(f"ride {ride_count} and the current fare :₹{current_fare}")

    print("-------------------------------")
  elif choice==0:
    print(f"Total fare :{fare}")
    break
  else:
    print("❌❌Give the wrong choice❌❌")

```
