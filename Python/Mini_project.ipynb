<H4>Project Title</H4>
<H4>Create My Diary App</H4>

Objective
Design a Python-based secure diary application where users can store their personal notes in an encrypted format.
The program must authenticate the user using a username and secret key before allowing access to the diary file.
The diary data should always remain encrypted inside the file to ensure privacy.
 When the user reads the diary, the application must decrypt the content before displaying it.

Functional Requirements

1. User Authentication
When the program starts, it should ask the user for:
Username


Secret Key


The program should check whether the corresponding diary file exists.
Example file naming format:
username_diary.txt
Example:
amar_diary.txt
If the file exists, the user is allowed to proceed.
If the file does not exist, display the following message:
Diary not found for this user.

2. Menu Options
After successful authentication, the program should display a menu:
1. Read Diary
2. Write Diary
3. Exit
The user must select one option to continue.

3. Read Diary
If the user selects Read Diary, the program should:
Open the diary file.


Read the encrypted data stored in the file.


Decrypt the content using the secret key.


Display the decrypted diary entries to the user.


Example Output:
Your Diary Entries:

Today I started learning Python.
It was fun and interesting.

4. Write Diary
If the user selects Write Diary, the program should:
Ask the user to enter a diary entry.


Encrypt the entry using the secret key.


Append the encrypted text to the diary file.


Example Interaction:
Enter your diary entry:
Today I completed my Python mini project.

Entry saved successfully.
The stored content in the file will appear encrypted, for example:
Wrgdb#L#frpsohwhg#p|#S|wkrq#plql#surmhfw

5. Exit Option
If the user selects Exit, the program should terminate gracefully.
Example:
Thank you for using My Diary App.
Goodbye!

Encryption Requirement
The program should implement a simple encryption technique, such as:
Caesar Cipher (Character Shifting)
Each character should be shifted based on the secret key value.
Example:
Original Text : hello
Encrypted Text : khoor
The same secret key must be used to decrypt the content when reading the diary.

Example: Encryption and Decryption using Secret Key (XOR)
1. Encryption Code
import getpass

key = int(getpass.getpass("Enter secret key: "))
message = input("Enter message to encrypt: ")

encrypted_message = ""

for char in message:
   encrypted_message += chr(ord(char) ^ key)

print("Encrypted Message:", encrypted_message)
Sample Run

(The encrypted text will vary depending on the key.)

2. Decryption Code
The same key must be used to decrypt the message.
import getpass

key = int(getpass.getpass("Enter secret key: "))
encrypted_message = input("Enter encrypted message: ")

secret_message = ""

for char in encrypted_message:
   secret_message += chr(ord(char) ^ key)

print("Original Message is:", secret_message)

Sample Input / Output

Important Concept
XOR encryption works because:
A ^ B ^ B = A
Example:
Character → Encrypt with Key → Decrypt with Same Key → Original Character
So the same operation is used for both encryption and decryption.


```python
import getpass

log_data = {}

# --- HELPER FUNCTIONS ---
def process_text(text, key):
    result = ""
    for char in text:
        result += chr(ord(char) ^ key)
    return result

def write_diary(filename, key):
    message = input("Enter text to write in file:\n")
    encrypted_message = process_text(message + "\n", key)

    with open(filename, "a") as f:
        f.write(encrypted_message)
    print("Write successfully.")

def read_diary(filename, key):
    try:
        with open(filename, "r") as f:
            data = f.read()

        if data == "":
            print("Diary is empty.")
        else:
            decrypted_message = process_text(data, key)
            print(f"\n--- Decrypted Message ---\n{decrypted_message}")
            print("Got original message successfully.")
    except FileNotFoundError:
        print("File not found.")

# --- MAIN LOGIC FUNCTIONS ---

def create_account(user_name, passw):
    if user_name not in log_data:
        log_data[user_name] = passw
        # Create a blank file for the new user
        with open(f"{user_name}_diary.txt", "w") as f:
            pass
        print("Account created successfully!")
    else:
        print("already have account")

def inner_menu(user_name, key):
    filename = f"{user_name}_diary.txt"
    while True:
        print(f"\n--- {user_name.upper()}'S DIARY ---")
        print("1. Write")
        print("2. Read")
        print("3. Log out")

        sub_comm = int(input("Enter choice: "))
        if sub_comm == 1:
            write_diary(filename, key)
        elif sub_comm == 2:
            read_diary(filename, key)
        elif sub_comm == 3:
            print("Logged out.")
            break
        else:
            print("Wrong Input")

def login(user_name):
    if user_name in log_data:
        actual_key = log_data[user_name]
        cnt = 3

        # 3-Chances Logic
        while cnt > 0:
            entered_key = int(getpass.getpass("Enter your password/key: "))
            if entered_key == actual_key:
                print("Login successful!")
                inner_menu(user_name, actual_key) # Go to the diary menu!
                break
            else:
                cnt -= 1
                print(f"Wrong key! You have {cnt} chances left.")

        if cnt == 0:
            print("system is failed down due to wrong pincode attempts")
    else:
        print("User not found. Please create an account.")


print("Our Private Diary".center(40,"-"))
while True:
    print("".center(40,"-"))

    print('''1. log in
2. create account
3. Exit''')

    comm = int(input("\nEnter your Command: "))

    if comm == 1:
        user_name = input("user_name: ")
        login(user_name)

    elif comm == 2:
        user_name = input("user_name: ")
        passw = int(getpass.getpass("Enter your password/key (numbers only): "))
        create_account(user_name, passw)

    elif comm == 3:
        print("Exiting program...")
        break

    else:
        print("Wrong Input")
```
