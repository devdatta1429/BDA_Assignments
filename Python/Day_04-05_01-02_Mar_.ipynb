```python
#Shallow Copy

data=[1,2,3]
d1=data

print(data)  # [1, 2, 3]
print(d1)    # [1, 2, 3]

d1[0]=5
print(data)  # [5, 2, 3]
print(d1)    # [5, 2, 3]

print(id(data)) # 132352657582272
print(id(d1))   # 132352657582272
```

```python
#Deep Copy

data=[1,2,3]
d1=data.copy()
d1[0]=5
print(data)  # [1, 2, 3]
print(d1)    # [1, 2, 3]

print(id(data)) # 132352657583424
print(id(d1))   # 132352657583232
```

```python
menu_items={"Tea":10, "Coffee": 20, "Water":20, "BreadToast":25}
user_cart={}

def menu():
    i=1
    for key,value in menu_items.items():
        print(f"{i}. {key} {value}.00 ₹")
        i+=1
    print("---------------------")

def cart(name,quan):
    if name in menu_items:
        user_cart[name] = quan
        print("---------------------")
        print(f"+ Added {quan} {name}")
        print("---------------------")
    else:
        print("! Item not on menu.")

def bill():
    total=0
    for key,value in user_cart.items():
        # print(f"{key} : {value}")

        price= menu_items[key]

        subtotal = price * value
        print("Items  Price  Quan  Total")
        print(f" {key}:   {price}  X  {value}  :  {subtotal}")
        total+=subtotal
    print(f"Total : ₹ {total}")
    print("---------------------")


while True:
    print("1. Order  2. Bill  3. Exit")
    comm=int(input("Choice :"))
    print("---------------------")

    if comm==1:
        menu()
        name=input("Enter Item Name :")
        name=name.strip().title()
        quan=int(input("Enter Quantity :"))
        cart(name,quan)
    elif comm==2:
        if not user_cart:
            print("Cart is empty")
        else:
            bill()
    elif comm==3:
        break
    else:
        print("Wrong input")

```
