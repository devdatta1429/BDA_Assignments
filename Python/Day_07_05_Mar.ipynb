```python
class Error(Exception):
    def __init__(self,except1):
        self.except1=except1
    def __str__(self):
        return f"Exception raised: at {self.except1} error for file not found"

class error1(Exception):
    def __init__(self,except1):
        self.except1=except1
    def __str__(self):
        return f"Exception raised: at {self.except1} error for user not found"
```

```python
try:
    print("welcome")
    code = int(input("enter ur number"))
    if code==404:
        raise Error(code)
    elif code==500:
        raise error1(code)
    else:
        print("wrong code")
except Exception as msg:
    print("Error:",msg)
```
