```python
### Gemini version


def combine_files(file1, file2, output_file="combine.txt"):
    # Open both source files
    with open(file1, 'r') as f1, open(file2, 'r') as f2:
        lines1 = f1.readlines()
        lines2 = f2.readlines()

    # Write alternately into the output file
    with open(output_file, 'w') as out:
        max_len = max(len(lines1), len(lines2))
        for i in range(max_len):
            if i < len(lines1):
                out.write(lines1[i].strip() + "\n")
            if i < len(lines2):
                out.write(lines2[i].strip() + "\n")

    print(f"Contents combined into {output_file}")

# Example usage:
file1 = input("Enter first filename: ")
file2 = input("Enter second filename: ")
combine_files(file1, file2)
```

<H5>#Write a code to accept two source filenames from the user and create a new file called combine.txt, which will contain the contents of both the files in an alternate manner.

```python

file1 = input("Enter first filename: ")
file2 = input("Enter second filename: ")

try:
    f1 = open(file1, "r")
    f2 = open(file2, "r")
    fw = open("combine.txt", "w")

    while True:
        line1 = f1.readline()
        line2 = f2.readline()

        if line1 == "" and line2 == "":
            break

        if line1 != "":
            fw.write(line1)

        if line2 != "":
            fw.write(line2)

    print("Files combined successfully into combine.txt")

    f1.close()
    f2.close()
    fw.close()

except FileNotFoundError:
    print("One of the files was not found")
```
