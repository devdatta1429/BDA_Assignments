3️⃣ Petrol Expense Tracker
Problem Definition:
You are planning a road trip. Calculate total petrol cost based on distance, mileage, and petrol price.

Arithmetic Operators Used: /, *, +

Formula:
Petrol Needed = Distance / Mileage
Total Cost = Petrol Needed * Petrol Price

Sample Input:
Distance = 300 km
Mileage = 15 km/litre
Petrol Price = 110 per litre

Sample Output:
Petrol Needed = 20 litres
Total Cost = 2200

```python
distance= float(input("Enter your distance: "))
mileage = float(input("Enter your mileage: "))
price = float(input("enter the fuel price: "))

fuel_req = distance/mileage
total_money = fuel_req * price
print(f"Total money required {total_money}")
```
