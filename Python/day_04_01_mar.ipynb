```python
#Shallow Copy

data=[1,2,3]
d1=data

print(data)  # [1, 2, 3]
print(d1)    # [1, 2, 3]

d1[0]=5
print(data)  # [5, 2, 3]
print(d1)    # [5, 2, 3]

print(id(data)) # 132352657582272
print(id(d1))   # 132352657582272
```

```python
#Deep Copy

data=[1,2,3]
d1=data.copy()
d1[0]=5
print(data)  # [1, 2, 3]
print(d1)    # [1, 2, 3]

print(id(data)) # 132352657583424
print(id(d1))   # 132352657583232
```

```python
def smaller(num=1):
  data=[10,20,30,25,35,45,65,50,60,80,70]
  return sorted(data)[num-1]

smaller(0)

def larger(num):
  data=[10,20,30,25,35,45,65,50,60,80,70]
  return sorted(data)[-(num)]

larger(1)
```
