#### Q3. Write a program to find the factorial of a number.

```python
def facto(num):
    if num==1 or num==0:
        return 1
    return num*facto(num-1)

facto(5)
```

#### Q4. Write a program to check if a string is a palindrome.

```python
str1= 'acjajca'

print(str1==str1[::-1])
#-------------------------------------------------
str2=''
for char in str1:
    str2 = char + str2
print(str1==str2)
#-------------------------------------------------
left  = 0
right = len(str1)-1

is_pallindrome = True

while left < right :
    if str1[left] != str1[right]:
        is_pallindrome = False
        break

    left+=1
    right-=1

print(is_pallindrome)
```

#### Q5. Write a program to find the largest element in an array.

```python
lst = [2,3,6,4,5,7,45,87,65,8,9,5,14,34,12]

print(max(lst))

max_num = 0
for i in lst:
    if max_num < i:
        max_num = i
print(max_num)

#------------------------  2nd Element -------------------

print(sorted(lst)[-2])

max_num = 0
max_2nd = 0

for i in lst:
    if max_num < i :
        max_2nd = max_num
        max_num = i
    elif max_2nd < i:
        max_2nd = i

print(max_2nd)
```

#### Q7. Write a program to check if a number is prime.

```python
def prime(num):
    if num == 1 or num == 0:
        return 'not a prime'
    for i in range(2,num):
        if num%i==0:
            return "not Prime"
    return 'Prime'

prime(1)

```
