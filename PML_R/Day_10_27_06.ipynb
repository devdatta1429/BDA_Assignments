# 🔰RAG🔰

# 🔰 Large Language Models (LLMs)

There are **4 common ways to build AI-powered applications/websites using LLMs:**

## 1. Prompt Engineering

Prompt Engineering is the process of designing effective prompts (instructions) to get the desired output from an LLM.

### Advantages
- Easy to implement
- No training required
- Fast development
- Cost-effective

### Limitations
- Limited by the model's knowledge cutoff
- Can produce hallucinations
- Difficult for domain-specific knowledge

---

# 2. RAG (Retrieval-Augmented Generation) 🔰

RAG combines an LLM with an external knowledge source.

Instead of relying only on the model's internal knowledge, it first retrieves relevant information from documents, databases, or websites and then generates the answer.

## Why RAG?

### ✅ Overcomes Knowledge Cutoff
- Accesses the latest information.
- Uses external documents instead of relying only on the pretrained model.

### ✅ Reduces Hallucinations
- Answers are generated using retrieved context.
- Produces more factual and reliable responses.

### ✅ Domain-Specific Applications
Useful for:
- Company documents
- Healthcare
- Legal systems
- Banking
- Education
- Customer support

### RAG Workflow

```
User Question
      │
      ▼
Retriever (Vector Database)
      │
Retrieve Relevant Documents
      │
      ▼
LLM
      │
Generate Answer using Retrieved Context
      │
      ▼
Final Response
```

## Components of RAG

1. Documents
2. Embedding Model
3. Vector Database
4. Retriever
5. Large Language Model (LLM)

## Popular Vector Databases

- FAISS
- ChromaDB
- Pinecone
- Weaviate
- Milvus

## Advantages

- Latest information
- Reduces hallucinations
- No need for retraining
- Easy to update knowledge
- Works well with private/company data

## Disadvantages

- Retrieval quality affects performance
- More complex architecture
- Slightly slower than direct prompting

---

# 3. Fine-Tuning

Fine-tuning means training a pretrained LLM on your own dataset so it learns a specific task or domain.

## When to Use

- Domain-specific language
- Custom writing style
- Classification tasks
- Code generation
- Medical or legal applications

## Advantages

- Better performance on specialized tasks
- Learns organization-specific behavior
- More consistent responses

## Disadvantages

- Expensive
- Requires labeled data
- Time-consuming
- Needs GPUs for training

---

# 4. AI Agents (Tool Calling / Function Calling)

AI Agents are LLM-powered systems that can reason, plan, and use external tools to complete tasks.

## Examples of Tools

- Web Search
- Calculator
- Database
- APIs
- Python
- Email
- Calendar

## Workflow

```
User Request
      │
      ▼
LLM Agent
      │
Chooses Tool
      │
Calls Tool/API
      │
Gets Result
      │
      ▼
Final Answer
```

## Advantages

- Can perform multi-step tasks
- Uses external tools
- Can automate workflows
- More powerful than simple chatbots

## Popular Frameworks

- LangChain
- LangGraph
- LlamaIndex
- CrewAI
- AutoGen

# 🔰 Retrieval-Augmented Generation (RAG)

## Problems with Plain LLMs

A standalone LLM has several limitations:

- ❌ Knowledge cutoff (cannot access recent information)
- ❌ Hallucinations (may generate incorrect facts)
- ❌ No access to private/company data
- ❌ Limited domain-specific knowledge
- ❌ Cannot verify answers from external sources

---

# What RAG Fixes

RAG improves LLMs by retrieving relevant information before generating an answer.

### Benefits

- ✅ Accesses the latest information
- ✅ Reduces hallucinations
- ✅ Uses private/company documents
- ✅ Produces more accurate responses
- ✅ Easy to update knowledge without retraining

---

# What is RAG?

**RAG (Retrieval-Augmented Generation)** is a technique that combines:

- **Retrieval** → Search for relevant information
- **Generation** → Generate an answer using the retrieved information

Instead of answering only from its internal knowledge, the model first retrieves relevant documents and then generates the response.

---

# RAG Workflow

### Step 1
👤 User asks a question.

### Step 2
🔍 Retriever searches the knowledge base or vector database.

### Step 3
📄 Relevant context is added to the prompt.

### Step 4
🤖 Generator (LLM/SLM) produces the final answer using the retrieved context.

```
User Question
      │
      ▼
Retriever
      │
Retrieve Relevant Documents
      │
      ▼
Add Context to Prompt
      │
      ▼
LLM / SLM
      │
Generate Answer
      │
      ▼
Final Response
```

---

# Simple Analogy

### Plain LLM

📖 **Open-book exam without the book**

The model answers only from what it remembers.

### RAG

📚 **Open-book exam with the correct pages already open**

The model first finds the right information and then answers.

---

# How RAG Works

**RAG = Retrieval + Generation**

The retrieval system finds useful information, and the generator creates the final response using that information.

---

# RAG Architecture

A complete RAG pipeline usually contains:

- Retrieval System
- Vector Database
- Re-ranker (optional)
- Generative Model (LLM/SLM)

```
User Query
      │
      ▼
Retriever
      │
      ▼
Vector Database
      │
Retrieve Top Documents
      │
      ▼
Re-ranker (Optional)
      │
Select Best Documents
      │
      ▼
LLM / SLM
      │
Generate Final Answer
```

---

# Vector Database

A **Vector Database** stores document embeddings (vector representations).

Instead of searching by exact words, it searches by **semantic similarity** (meaning).

### Popular Vector Databases

- FAISS
- ChromaDB
- Pinecone
- Milvus
- Weaviate

---

# Three Phases of RAG

## Phase 1: Retrieval (Search)

- User asks a question.
- Retriever searches the vector database.
- Finds the most relevant documents.

---

## Phase 2: Augmentation (Context Injection)

- Retrieved documents are added to the prompt.
- The LLM receives both the question and the retrieved context.

---

## Phase 3: Generation (Answer)

- The LLM uses the retrieved context.
- Generates a factual and context-aware response.

---

# Two Main Components

## 1. Retriever

Responsibilities:

- Searches the vector database
- Finds relevant documents
- Returns the top matching results

Examples:

- Dense Retriever
- BM25
- Hybrid Search

---

## 2. Generator

Responsibilities:

- Reads the retrieved context
- Understands the user's question
- Generates the final answer

Examples:

- GPT
- Llama
- Gemini
- Mistral

---

# Complete RAG Pipeline

```
Documents
     │
     ▼
Text Chunking
     │
     ▼
Embedding Model
     │
     ▼
Vector Database
     │
     ▼
User Question
     │
     ▼
Retriever
     │
Retrieve Relevant Chunks
     │
     ▼
Context Injection
     │
     ▼
LLM / SLM
     │
Generate Answer
     │
     ▼
Final Response
```

---

# Advantages of RAG

- Latest information
- Lower hallucination rate
- Supports private knowledge bases
- No retraining required
- Cost-effective
- Easy to update documents

---

# Limitations of RAG

- Retrieval quality affects accuracy
- More complex architecture
- Slightly higher response time
- Requires a vector database

---

# Key Takeaways

- **Plain LLM** → Answers from pretrained knowledge only.
- **RAG** → Retrieves relevant information before generating an answer.
- **Retriever** → Finds relevant documents.
- **Generator** → Produces the final answer.
- **Vector Database** → Stores document embeddings for semantic search.
- **RAG = Retrieval + Augmented Generation** for more accurate and up-to-date responses.
