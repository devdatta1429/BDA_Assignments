# Natural Language Processing (NLP)

## What is NLP?

Natural Language Processing (NLP) is a branch of Artificial Intelligence (AI) that enables computers to understand, interpret, process, and generate human language.

### Applications of NLP

1. Spell Checking
   - Example: Microsoft Word, Grammarly

2. Keyword Searching
   - Example: Google Search Engine

3. Information Extraction
   - Extracting names, dates, locations, etc. from documents

4. Advertisement Matching
   - Showing relevant advertisements based on user searches

5. Chatbots and Virtual Assistants
   - ChatGPT, Siri, Alexa

6. Machine Translation
   - Google Translate

7. Sentiment Analysis
   - Product review analysis

# Why is NLP Hard?

Human languages are easy for humans but difficult for machines.

Challenges:

1. Ambiguity

Example:

```text
I saw a man with a telescope.
```

Who has the telescope?

2. Synonyms

```text
Big = Large
```

Same meaning, different words.

3. Context Understanding

```text
Apple
```

Can refer to:
- Fruit
- Company

4. Sarcasm

```text
Great! Another exam tomorrow.
```

Actual sentiment may be negative.

### Conclusion

English (or any natural language) is not naturally understood by machines. Therefore NLP techniques are required.

# Document-Term Matrix (DTM)

A matrix representing the frequency of terms in documents.

Example:

| Word | Doc1 | Doc2 |
|--------|--------|--------|
| NLP | 1 | 1 |
| Love | 1 | 0 |
| Interesting | 0 | 1 |

Rows → Terms

Columns → Documents

Values → Word Frequencies

# WordNet

WordNet is a lexical database (dictionary-like database) of English words.

It stores:

- Synonyms
- Antonyms
- Meanings
- Relationships between words

Example:

```text

|
|-- Glad
|-- Joyful
|-- Cheerful
```

WordNet is available through NLTK.

# Processing the Corpus

Typical NLP preprocessing pipeline:

## 1. Lowercasing

Convert all text to lowercase.

Example:

```text
NLP
```

becomes

```text
nlp
```

---

## 2. Tokenization

Split text into smaller units called tokens.

Example:

```text
I love NLP
```

Tokens:

```python
["I", "love", "NLP"]
```

---

## 3. Removing Stop Words

Stop words are common words that carry little meaning.

Examples:

```text
is
am
the
a
an
of
```

After removal:

```text
love NLP
```

---

## 4. Stemming / Lemmatization

Convert words to their root form.

---

## 5. Removing Punctuation

Remove symbols such as:

```text
.
,
!
?
:
;
```

Example:

```text
Hello!!!
```

becomes

```text
Hello
```

# Tokenization

Tokenization is the process of splitting text into tokens.

A single token is called a:

```text
Unigram
```

---

## Types of Tokens

### Unigram

Single word

Example:

```text
I
love
NLP
```

---

### Bigram

Two consecutive words

Example:

```text
I love
love NLP
```

---

### Trigram

Three consecutive words

Example:

```text
I love NLP
```

---

### N-Gram

General form containing N words.

Examples:

- 1 word → Unigram
- 2 words → Bigram
- 3 words → Trigram
- N words → N-Gram

---

# Advantages of Tokenization

### 1. Reduces Search Space

Searching tokens is easier than searching full documents.

### 2. Reduces Storage Complexity

Efficient indexing and retrieval.

---

# Stemming

Stemming reduces words to their root form.

Example:

```text
Affecting
Affected
Affection
```

becomes

```text
Affect
```

The resulting word may or may not be a valid dictionary word.

---

# Lemmatization

Lemmatization converts words to their dictionary (base) form called a Lemma.

Example:

```text
Running → Run
Better → Good
Happiness → Happy
```

Produces meaningful words.

### Definition

Lemmatization is a text processing technique that reduces words to their base or dictionary form known as the lemma.

---

# Stemming Algorithms

## 1. Porter Stemmer

Most commonly used stemmer.

Example:

```text
Playing → Play
```

---

## 2. Snowball Stemmer

Improved version of Porter Stemmer.

Supports multiple languages.

---

## 3. Lancaster Stemmer

More aggressive stemming.

Example:

```text
Maximum → Max
```

---

## 4. Regex-Based Stemmer

Uses regular expressions to remove suffixes.

---

# Stemming vs Lemmatization

| Stemming | Lemmatization |
|-----------|-----------|
| Rule Based | Dictionary Based |
| Faster | Slower |
| Less Accurate | More Accurate |
| May generate invalid words | Produces valid words |
| No context understanding | Considers context |

### Example

Stemming:

```text
Happiness → Happi
```

Lemmatization:

```text
Happiness → Happy
```
