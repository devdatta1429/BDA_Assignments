# 🔰 Word Embeddings 🔰

## What is Word Embedding?

Word Embedding converts words into dense numerical vectors while preserving semantic meaning.

### Example (Word2Vec)

```text
King - Man + Woman = Queen
````

or

```text
King : Man :: Woman : Queen
```

Words with similar meanings have similar vector representations.

---

## 🎯 Uses of Embeddings

1. Capture Semantic Meaning
2. Dimensionality Reduction
3. Context Awareness
4. Improve Model Performance

### Applications

* Text Classification
* Sentiment Analysis
* Machine Translation
* Question Answering
* Text Generation

---

# Types of Embeddings

## 1. Traditional Vectorization Techniques

* One-Hot Encoding
* Bag of Words (BoW)
* TF-IDF

### Limitation

* Sparse vectors
* No semantic understanding
* High dimensionality

---

## 2. Static Word Embeddings

Same word → Same vector everywhere

### Techniques

* Word2Vec
* GloVe
* FastText

### Example

```text
Bank (river)
Bank (money)

→ Same Vector
```

### Advantages

* Dense representation
* Captures semantic relationships
* Computationally efficient

---

## 3. Contextual Word Embeddings

Different contexts → Different vectors

### Techniques

* ELMo
* BERT (Bidirectional Encoder Representations from Transformers)
* RoBERTa
* ALBERT
* DistilBERT
* GPT
* T5
* BART

### Example

```text
I deposited money in the bank.
I sat near the river bank.
```

```text
Bank → Vector A
Bank → Vector B
```

### Advantages

* Context aware
* Better understanding of language
* Improved NLP performance

---

# Static vs Contextual Embeddings

| Feature                      | Static        | Contextual          |
| ---------------------------- | ------------- | ------------------- |
| Context Awareness            | ❌ No          | ✅ Yes               |
| Same Word, Different Meaning | ❌ Same Vector | ✅ Different Vectors |
| OOV Handling                 | Limited       | Better              |
| Computation Cost             | Low           | High                |
| Accuracy                     | Moderate      | High                |

---

## 4. Sentence & Document Embeddings

Represent an entire sentence or document as a single vector.

### Examples

* Sentence-BERT (SBERT)
* Doc2Vec
* Universal Sentence Encoder (USE)

### Applications

* Semantic Search
* Document Similarity
* Recommendation Systems

---

## 5. Multimodal Embeddings

Combine multiple data types into a shared embedding space.

### Modalities

* Text
* Image
* Audio
* Video

### Examples

* CLIP
* Flamingo
* Gemini-like Models

### Applications

* Image Captioning
* Visual Question Answering
* Cross-Modal Retrieval


# 🔰 RNN (Recurrent Neural Network) 🔰

### Why RNN instead of ANN?

**ANN**
- Fixed-size input
- No memory of previous inputs
- Treats each input independently

**Problem**
- Text data is sequential
- Input length can vary
- Word order matters

**RNN**
- Handles variable-length sequences
- Maintains information from previous inputs
- Captures context and dependencies between words

### Example

Sentence:

I love Machine Learning

```text
I  → h₁
      ↓
love → h₂
      ↓
Machine → h₃
      ↓
Learning → Output
````

### Conclusion

* ANN → Fixed-size input, no memory
* RNN → Sequential input, remembers previous information
* Better for NLP and text classification tasks
