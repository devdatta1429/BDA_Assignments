```python
text = "NLP is one of the most exciting fields of artificial intelligence"

tokens = word_tokenize(text.lower())
print('tokens :', tokens)
stop_word = set(stopwords.words('english'))

filtered=[]
for word in tokens:
    if word not in stop_word:
        filtered.append(word)

print('filtered', filtered)
```

```python
nlp = spacy.load('en_core_web_sm')
nlp

doc=nlp(text)
doc
```

```python
doc = nlp(text)
filtered_spacy = [t.text for t in doc if not t.is_stop and not t.is_punct]
print("spaCy filtered:", filtered_spacy)
# ['NLP', 'one', 'exciting', 'fields', 'artificial', 'intelligence']

# ── Adding/removing custom stopwords ────────
nlp.vocab["nlp"].is_stop = True   # add
nlp.vocab["not"].is_stop = False  # remove (critical for sentiment analysis!)
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
stem1 = PorterStemmer()
```

```python
words=['run','running','ran','runs','easily','fairly','happy','affection','caring']
for word in words:
    print(word+'---->'+stem1.stem(word))
```

```python
from nltk.stem import WordNetLemmatizer
lem1 = WordNetLemmatizer()
```

```python
words=['run','running','ran','runs','easily','fairly','happy','affection','caring']
for word in words:
    print(f"{'stem :':<5}",word+'---->'+stem1.stem(word))
    print(f"{'lem :':<5}",word+'---->'+lem1.lemmatize(word))
    print()
```
