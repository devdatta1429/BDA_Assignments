# 🔰 POS Tags (Part of Speech) in NLTK

## What is POS Tagging?

POS (Part-of-Speech) Tagging is the process of assigning grammatical labels to each word in a sentence.

Example:

Sentence:
```python
"I love machine learning"
```

POS Tagged:
```python
[('I', 'PRP'),
 ('love', 'VBP'),
 ('machine', 'NN'),
 ('learning', 'NN')]
```

---

## Why POS Tagging is Important?

- Helps understand sentence structure
- Improves NLP tasks:
  - Text Classification
  - Named Entity Recognition (NER)
  - Machine Translation
  - Sentiment Analysis
  - Information Extraction
- Used as features in Deep Learning models

---

# 🔰 Important POS Tags: Verbs, Adjectives, and Adverbs

## 1️⃣ Verb Tags (VB*)

Verbs represent actions, events, or states.

| Tag | Meaning | Example |
|------|---------|---------|
| VB | Base Form Verb | go, eat, run |
| VBD | Past Tense Verb | went, ate, ran |
| VBG | Gerund/Present Participle | going, eating, running |
| VBN | Past Participle | gone, eaten, written |
| VBP | Present Tense (Non-3rd Person) | go, eat |
| VBZ | Present Tense (3rd Person Singular) | goes, eats |

### Examples

```python
go      -> VB
went    -> VBD
going   -> VBG
gone    -> VBN
eat     -> VBP
eats    -> VBZ
```

### Sentence Example

```text
She goes to school.
```

```python
She   -> PRP
goes  -> VBZ
to    -> TO
school-> NN
```

---

## 2️⃣ Adjective Tags (JJ*)

Adjectives describe nouns or pronouns.

| Tag | Meaning | Example |
|------|---------|---------|

| JJR | Comparative Adjective | taller, better |
| JJS | Superlative Adjective | tallest, best |

### Examples

```python

```

## 3️⃣ Adverb Tags (RB*)

Adverbs modify verbs, adjectives, or other adverbs.

| Tag | Meaning | Example |
|------|---------|---------|
| RB | Adverb | quickly, slowly |
| RBR | Comparative Adverb | faster, harder |
| RBS | Superlative Adverb | fastest, hardest |

### Examples

```python

```

### Sentence Example

```text

```


# 🔰 Named Entity Recognition (NER)

## What is NER?

**Named Entity Recognition (NER)** is an NLP technique used to identify and classify important entities in text into predefined categories.

Examples of entities:

- Person Names
- Organizations
- Locations
- Dates
- Time
- Money
- Percentages

---

## Example

Sentence:

```text
Sundar Pichai is the CEO of Google and lives in California.
```

NER Output:

```text
Sundar Pichai → PERSON
Google        → ORGANIZATION
California    → LOCATION
```

---

## Common NER Tags

| Tag | Meaning | Example |
|------|---------|---------|
| PERSON | Person Name | John, Elon Musk |
| ORGANIZATION (ORG) | Company/Institution | Google, Microsoft |
| GPE | Country, City, State | India, Mumbai |
| LOCATION (LOC) | Geographic Location | Himalayas, Pacific Ocean |
| DATE | Date | 10 June 2025 |
| TIME | Time | 10:30 AM |
| MONEY | Monetary Value | $500 |
| PERCENT | Percentage | 75% |
| FACILITY (FAC) | Buildings/Airports | Airport, Stadium |
| EVENT | Named Event | Olympics |
| PRODUCT | Product Name | iPhone |
| LANGUAGE | Language | English, Hindi |

---

````python
import re
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')

nltk.download('averaged_perceptron_tagger_eng')#Pos tagging

nltk.download('maxent_ne_chunker_tab') #Chunking
nltk.download('words')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.chunk import ne_chunk
````

````python
text = "Ruchika is teaching Natural Language Processing in Mumbai using Python."

clean_text = text.lower()
clean_text = re.sub(r'[^\w\s]', '', clean_text)
print(clean_text)
````

````python
tokens = word_tokenize(clean_text)
print(tokens)
````

````python
stop_words = set(stopwords.words('english'))

filtered_tokens = [
    word for word in tokens
    if word not in stop_words
]
print(tokens)
print(filtered_tokens)
````

````python
stemmer = PorterStemmer()

stems = [stemmer.stem(word) for word in filtered_tokens]

for stem in stems:
  print(stem)

````

````python
lemmatizer = WordNetLemmatizer()

lemmas = [
    lemmatizer.lemmatize(word)
    for word in filtered_tokens
]

for lemma in lemmas:
    print(lemma)

````

````python
original_tokens = word_tokenize(text)

pos_tags = pos_tag(original_tokens)

for word, tag in pos_tags:
    print(word,tag)
````

Chunking(Shallow Parsing) Rule based grammar
Chunking identifies phrase structure (syntax), while NER identifies entity types (semantics).

````python

grammar = r"""
    NP: {<DT>?<JJ>*<NN.*>+}
    VP: {<VB.*><NP|PP>*}

"""
chunk_parser = nltk.RegexpParser(grammar)

chunk_tree = chunk_parser.parse(pos_tags)
print(chunk_tree)

````

# NER(input is always a pos_tag)

````python
ner_tree = ne_chunk(pos_tags)

print(ner_tree)
````

````python
import pandas as pd
import nltk
import spacy
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from collections import Counter

from sklearn.feature_extraction.text import CountVectorizer


nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
````

````python
from sklearn.feature_extraction.text import CountVectorizer
# Sample documents
documents = [
    "I love NLP",
    "NLP is interesting",
    "I love machine learning"
]

# Create BoW model
vectorizer = CountVectorizer() #creates vocabulary


# Convert text into BoW matrix
X = vectorizer.fit_transform(documents)#counts occurances based on vocabulary

# Display as table
df = pd.DataFrame(
    X.toarray(),#sparse matrix to normal
    columns=vectorizer.get_feature_names_out()
)

print(df)
````

TF:How often a word appears in a document\
IDF:Gives less importance to words appearing in many documents.

````python
from sklearn.feature_extraction.text import TfidfVectorizer


# Sample documents
documents = [
    "I love NLP",
    "NLP is interesting",
    "I love machine learning"
]

# Create TF-IDF model
vectorizer = TfidfVectorizer()

# Convert text into TF-IDF matrix
X = vectorizer.fit_transform(documents)

# Display as table
df = pd.DataFrame(
    X.toarray(),
    columns=vectorizer.get_feature_names_out()
)

print(df.round(2))
````
