# 🔰 Transformer Architecture

The Transformer model consists of two main components:

- **Encoder**
- **Decoder**

---

# 🔹 Encoder

The **Encoder** has **2 main layers**:

## 1. Multi-Head Self-Attention
- Allows the model to focus on different parts of the input sequence simultaneously.
- Captures relationships between words regardless of their positions.

## 2. Feed Forward Neural Network (FFN)
- A fully connected neural network.
- Applied independently to each token.
- Helps learn complex patterns from the attention output.

### Encoder Flow

```text
Input
   │
   ▼
Multi-Head Self-Attention
   │
   ▼
Feed Forward Network
   │
   ▼
Encoder Output
```

---

# 🔹 Decoder

The **Decoder** has **3 main layers**:

## 1. Masked Multi-Head Self-Attention
- Similar to self-attention.
- Uses masking so the model cannot see future words during training.
- Ensures predictions are generated one token at a time.

## 2. Encoder-Decoder Cross Attention
- The decoder attends to the encoder's output.
- Helps the decoder understand the input sequence while generating output.

## 3. Feed Forward Neural Network (FFN)
- Processes the attention output.
- Learns higher-level representations.

### Decoder Flow

```text
Previous Output
       │
       ▼
Masked Multi-Head Attention
       │
       ▼
Encoder-Decoder Cross Attention
       │
       ▼
Feed Forward Network
       │
       ▼
Final Output
```

---

# 🔰 BERT

**BERT** stands for:

**B**idirectional **E**ncoder **R**epresentations from **T**ransformers

## Key Points

- Uses **only the Encoder** part of the Transformer.
- Reads text in **both directions** (left-to-right and right-to-left).
- Learns context from surrounding words.
- Pre-trained on large text datasets.

### Features

- ✅ Bidirectional Context
- ✅ Deep Language Understanding
- ✅ Context-Aware Word Representations
- ✅ Excellent for NLP tasks

---

# Applications of BERT

- Text Classification
- Sentiment Analysis
- Question Answering
- Named Entity Recognition (NER)
- Text Summarization
- Search Engines
- Chatbots

---

# Quick Comparison

| Encoder | Decoder |
|----------|----------|
| Multi-Head Self-Attention | Masked Multi-Head Self-Attention |
| Feed Forward Network | Encoder-Decoder Cross Attention |
| Produces Context Representation | Generates Output Sequence |
| Used in BERT | Used in GPT & Translation Models |

---

# Summary

- **Encoder**
  - Multi-Head Self-Attention
  - Feed Forward Network

- **Decoder**
  - Masked Multi-Head Self-Attention
  - Encoder-Decoder Cross Attention
  - Feed Forward Network

- **BERT**
  - Uses only the **Encoder**
  - Bidirectional language model
  - Designed for language understanding tasks

# 🔰 BERT Model Architecture

**BERT** stands for:

**B**idirectional **E**ncoder **R**epresentations from **T**ransformers

BERT is an **encoder-only Transformer model** designed for understanding the context of text in both directions (left-to-right and right-to-left).

---

# Stage 1: Tokenization

The input sentence is first converted into tokens.

### Special Tokens

- **[CLS]** → Added at the **beginning** of every input sequence.
  - Used for sentence-level classification tasks.
- **[SEP]** → Used as a **separator** between two sentences or to mark the end of a sentence.

### Example

```text
Input:
I love NLP.

After Tokenization:
[CLS] I love NLP . [SEP]
```

For sentence-pair tasks:

```text
[CLS] Sentence A [SEP] Sentence B [SEP]
```

---

# Stage 2: Embedding Layer

Each input token is represented by combining three types of embeddings.

## 1. Token Embedding
- Represents the meaning of each word/token.
- Learned during training.

## 2. Segment Embedding
- Identifies which sentence a token belongs to.
- Used for sentence-pair tasks.

Example:

```text
Sentence A → Segment 0
Sentence B → Segment 1
```

## 3. Positional Embedding
- Provides information about the position of each token.
- Since Transformers do not process sequences sequentially, positional information is added.

---

## Final Input Embedding

```text
Final Embedding =
Token Embedding
+ Segment Embedding
+ Positional Embedding
```

```text
Token Embedding
        +
Segment Embedding
        +
Positional Embedding
        │
        ▼
Summed Embedding
```

---

# Stage 3: Encoder

BERT uses **only Transformer Encoders**.

Each encoder block contains:

## 1. Multi-Head Self-Attention
- Allows every word to attend to every other word.
- Captures contextual relationships.

## 2. Add & Layer Normalization
- Adds residual connections.
- Normalizes outputs for stable training.

## 3. Feed Forward Neural Network (FFN)
- Fully connected neural network.
- Learns higher-level representations.

## 4. Add & Layer Normalization
- Applied again after the FFN.

---

### Encoder Block

```text
Input
  │
  ▼
Multi-Head Self-Attention
  │
  ▼
Add & Layer Normalization
  │
  ▼
Feed Forward Network
  │
  ▼
Add & Layer Normalization
  │
  ▼
Output
```

---

# Stage 4: Contextual Output

The encoder produces **contextualized representations** for every token.

Each token is converted into a **context-aware vector**.

```text
Input Tokens
     │
     ▼
Encoder
     │
     ▼
Token Vectors
(Contextual Embeddings)
```

These vectors are used for downstream NLP tasks such as:

- Text Classification
- Sentiment Analysis
- Question Answering
- Named Entity Recognition (NER)
- Text Similarity

# 🔰 Multi-Head Self-Attention (Encoder)

The **Multi-Head Self-Attention** mechanism allows every input token to attend to all other tokens in the sequence and learn contextual relationships.

---

# Step 1: Input Embeddings (N Tokens)

The input sentence is converted into embeddings.

```text
Input Tokens (N Tokens)

Token₁  Token₂  Token₃ ... Tokenₙ
      │
      ▼
Input Embeddings
```

---

# Step 2: Create Query (Q), Key (K), and Value (V)

Each input embedding is transformed into three vectors:

- **Query (Q)** → What information am I looking for?
- **Key (K)** → What information do I contain?
- **Value (V)** → The actual information to pass forward.

```text
Input Embeddings
        │
 ┌──────┼──────┐
 ▼      ▼      ▼
Query   Key   Value
(Q)     (K)    (V)
```

---

# Step 3: Compute Attention Scores

The similarity between **Query** and **Key** is calculated.

### Formula

$$
\text{Attention Score} =
\frac{QK^T}{\sqrt{d_k}}
$$

where:

- **Q** = Query Matrix
- **K** = Key Matrix
- **dₖ** = Dimension of the Key vectors

---

# Step 4: Apply Softmax

Softmax converts the attention scores into probabilities.

### Formula

$$
\text{Attention Weights} =
\text{Softmax}
\left(
\frac{QK^T}{\sqrt{d_k}}
\right)
$$

These weights indicate how much attention each token should pay to every other token.

---

# Step 5: Weighted Sum of Values

The attention weights are multiplied by the **Value (V)** matrix.

### Formula

$$
\text{Output}=
\text{Softmax}
\left(
\frac{QK^T}{\sqrt{d_k}}
\right)V
$$

This produces the attention output.

---

# Step 6: Multi-Head Attention

Instead of using one attention mechanism, the Transformer uses multiple attention heads.

```text
Head 1
Head 2
Head 3
...
Head h
```

Each head learns different relationships in the sentence.

---

# Step 7: Concatenate + Linear Projection

Outputs from all attention heads are:

- Concatenated
- Passed through a Linear layer

```text
Head1
Head2
Head3
...
Headh
   │
   ▼
Concatenate
   │
   ▼
Linear Projection
```

---

# Step 8: Add & Layer Normalization

A residual (skip) connection is added, followed by Layer Normalization.

```text
Attention Output
       │
       ▼
Residual Connection (+ Input)
       │
       ▼
Layer Normalization
```

This improves training stability and helps preserve information.

---

# Complete Multi-Head Self-Attention Flow

```text
Input Embeddings (N Tokens)
          │
          ▼
Generate Q, K, V
          │
          ▼
Compute Attention Scores
      QKᵀ / √dₖ
          │
          ▼
Softmax
          │
          ▼
Weighted Sum with V
          │
          ▼
Multiple Attention Heads
          │
          ▼
Concatenate
          │
          ▼
Linear Projection
          │
          ▼
Add (Residual Connection)
          │
          ▼
Layer Normalization
          │
          ▼
Output to Feed Forward Network
```

---

# Key Formulas

### Attention Score

$$
\frac{QK^T}{\sqrt{d_k}}
$$

### Attention Weights

$$
\text{Softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)
$$

### Self-Attention Output

$$
\text{Softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V
$$

---

# Summary

- **Input Embeddings** → Convert tokens into vectors.
- **Q, K, V** → Generate Query, Key, and Value matrices.
- **Attention Scores** → Compute \(QK^T/\sqrt{d_k}\).
- **Softmax** → Convert scores into attention weights.
- **Weighted Sum** → Multiply attention weights with Value (V).
- **Multi-Head Attention** → Multiple heads learn different relationships.
- **Concatenate + Linear Projection** → Combine outputs from all heads.
- **Add & Layer Normalization** → Apply residual connection and normalize before passing to the Feed Forward Network.

# 🔰 BERT Learning Process

BERT is trained in **two phases**:

1. **Pre-training**
2. **Fine-tuning**

---

# Phase 1: Pre-training

In the **pre-training** phase, BERT learns general language understanding from a very large text corpus (such as books and Wikipedia).

The model is trained using self-supervised learning, which means it creates its own training labels from raw text.

## Objectives of Pre-training

### 1. Masked Language Model (MLM)

- Some words in the sentence are randomly replaced with the **`[MASK]`** token.
- BERT predicts the original masked word using the surrounding context.
- Since BERT looks at both the left and right context, it learns **bidirectional representations**.

**Example**

```text
Input:
I love [MASK].

Target:
NLP
```

---

### 2. Next Sentence Prediction (NSP)

- BERT is given two sentences.
- The model predicts whether the second sentence actually follows the first sentence.

**Example**

```text
Sentence A:
I love Natural Language Processing.

Sentence B:
It is used in chatbots.

Output:
IsNext = True
```

```text
Sentence A:
I love Natural Language Processing.

Sentence B:
The weather is sunny today.

Output:
IsNext = False
```

---

## Output of Pre-training

- Learns grammar and sentence structure.
- Understands word meanings.
- Captures contextual relationships between words.
- Produces a general-purpose language model.

---

# Phase 2: Fine-tuning

After pre-training, BERT is adapted to a **specific NLP task** using labeled data.

Only a small task-specific layer is added on top of the pre-trained BERT model.

---

## Fine-tuning Process

```text
Pre-trained BERT
        │
        ▼
Add Task-Specific Layer
        │
        ▼
Train on Labeled Dataset
        │
        ▼
Task-Specific Model
```

---

## Applications of Fine-tuning

- Text Classification
- Sentiment Analysis
- Question Answering
- Named Entity Recognition (NER)
- Text Similarity
- Spam Detection
- Language Inference

---

# Learning Pipeline

```text
Raw Text Corpus
       │
       ▼
Phase 1: Pre-training
(MLM + NSP)
       │
       ▼
Pre-trained BERT Model
       │
       ▼
Phase 2: Fine-tuning
(Task-Specific Dataset)
       │
       ▼
Final NLP Model
```

---

# Comparison

| Feature | Pre-training | Fine-tuning |
|----------|--------------|-------------|
| Dataset | Large unlabeled text corpus | Small labeled dataset |
| Goal | Learn general language understanding | Learn a specific NLP task |
| Training Type | Self-supervised | Supervised |
| Output | Pre-trained BERT model | Task-specific BERT model |

---

# Key Points

- **Phase 1: Pre-training**
  - Uses large unlabeled text data.
  - Learns language representations.
  - Main objectives:
    - **Masked Language Modeling (MLM)**
    - **Next Sentence Prediction (NSP)**

- **Phase 2: Fine-tuning**
  - Uses labeled data.
  - Adds a task-specific output layer.
  - Adapts BERT for downstream NLP tasks like classification, question answering, and NER.

# 🔰 Masked Language Modeling (MLM)

**Masked Language Modeling (MLM)** is one of the two pre-training objectives used in **BERT**.

The goal is to predict the **original word** that has been hidden (masked) in the input sentence.

---

# Step 1: Select 15% of Tokens

During pre-training, **15% of the input tokens** are randomly selected for prediction.

Example:

```text
Original Sentence:
I love Natural Language Processing.

Selected Token:
Natural
```

---

# Step 2: Apply the 80-10-10 Rule

For the selected **15% of tokens**, BERT applies the following strategy:

| Percentage | Action |
|------------|--------|
| **80%** | Replace the token with **`[MASK]`** |
| **10%** | Replace the token with a **random word** |
| **10%** | Keep the token **unchanged** |

---

## 80% → Replace with `[MASK]`

```text
Original:
I love Natural Language Processing.

Input:
I love [MASK] Language Processing.

Target:
Natural
```

---

## 10% → Replace with a Random Word

```text
Original:
I love Natural Language Processing.

Input:
I love Computer Language Processing.

Target:
Natural
```

Even though **Computer** is inserted, BERT must still predict **Natural**.

---

## 10% → Keep the Original Word

```text
Original:
I love Natural Language Processing.

Input:
I love Natural Language Processing.

Target:
Natural
```

The word remains unchanged, but BERT is still trained to predict the selected token.

---

# Step 3: Predict the Original Token

BERT uses the surrounding context to predict the **original selected word**, regardless of whether it was:

- Masked (`[MASK]`)
- Replaced with a random word
- Left unchanged

---

# Why the 80-10-10 Rule?

Using only `[MASK]` tokens during training would create a mismatch because the `[MASK]` token is not present during most real-world inference tasks.

The **80-10-10 rule** helps BERT learn to:
- Handle masked words.
- Be robust to unexpected or incorrect words.
- Predict words even when the original token remains unchanged.

---

# What BERT Learns from MLM

Through Masked Language Modeling, BERT learns:

- ✅ Bidirectional context (left and right context)
- ✅ Word meanings based on surrounding words
- ✅ Semantic relationships between words
- ✅ Grammatical structure
- ✅ Context-aware word representations

---

# MLM Workflow

```text
Input Sentence
      │
      ▼
Randomly Select 15% of Tokens
      │
      ▼
Apply 80-10-10 Rule
 ┌───────────────┬───────────────┬──────────────┐
 │ 80% [MASK]    │ 10% Random    │ 10% Original │
 └───────────────┴───────────────┴──────────────┘
      │
      ▼
BERT Predicts the Original Token
      │
      ▼
Learns Contextual Word Representations
```

---

# Key Points

- **15%** of input tokens are selected for prediction.
- Apply the **80-10-10 rule**:
  - **80%** → Replace with **`[MASK]`**
  - **10%** → Replace with a **random word**
  - **10%** → Keep the **original word**
- BERT always predicts the **original token**.
- MLM helps BERT learn **contextual, bidirectional language representations**.

# 🔰 Next Sentence Prediction (NSP)

**Next Sentence Prediction (NSP)** is one of the two pre-training tasks used in the original **BERT** model.

The goal is to determine whether the **second sentence logically follows the first sentence**.

---

# Purpose of NSP

NSP helps BERT learn:

- Relationships between sentences
- Sentence coherence
- Context across multiple sentences

This is useful for tasks such as:

- Question Answering
- Natural Language Inference (NLI)
- Text Pair Classification

---

# How NSP Works

During pre-training, BERT receives **two sentences** as input.

```text
Sentence A
        +
Sentence B
```

The model predicts whether:

- **IsNext** → Sentence B follows Sentence A.
- **NotNext** → Sentence B is a random sentence.

---

# Input Format

```text
[CLS] Sentence A [SEP] Sentence B [SEP]
```

Where:

- **[CLS]** → Classification token
- **[SEP]** → Separator token

---

# Positive Example (IsNext)

```text
Sentence A:
The sun was setting.

Sentence B:
The sky turned orange.

Output:
IsNext = True
```

Here, Sentence B naturally follows Sentence A.

---

# Negative Example (NotNext)

```text
Sentence A:
The sun was setting.

Sentence B:
I bought a new laptop yesterday.

Output:
NotNext = True
```

Here, Sentence B is unrelated to Sentence A.

---

# Training Process

1. Select two sentences.
2. With **50% probability**, choose the actual next sentence (**IsNext**).
3. With **50% probability**, choose a random sentence from the corpus (**NotNext**).
4. BERT predicts whether the second sentence is the true next sentence.

---

# NSP Workflow

```text
Sentence A
      +
Sentence B
      │
      ▼
[CLS] Sentence A [SEP] Sentence B [SEP]
      │
      ▼
BERT Encoder
      │
      ▼
[CLS] Output Vector
      │
      ▼
Binary Classifier
      │
      ├── IsNext
      └── NotNext
```

---

# Output Classes

| Label | Meaning |
|--------|---------|
| **IsNext** | Sentence B follows Sentence A |
| **NotNext** | Sentence B is a random sentence |

---

# Why NSP is Used

NSP enables BERT to:

- Understand relationships between sentences.
- Learn sentence order and coherence.
- Improve performance on sentence-pair tasks.

---

# Note

The **original BERT** uses both:

- **Masked Language Modeling (MLM)**
- **Next Sentence Prediction (NSP)**

However, many newer BERT variants (such as **RoBERTa**) remove the NSP objective because later research showed that strong performance can be achieved without it.

---

# Summary

- NSP is a **binary classification** task.
- Input format:
  ```text
  [CLS] Sentence A [SEP] Sentence B [SEP]
  ```
- **50%** of training examples are **IsNext**.
- **50%** are **NotNext**.
- The **[CLS]** token representation is used to predict the final label.
- NSP helps BERT learn **sentence relationships** and **document-level context**.

# 🔰 Fine-Tuning Phase

After **pre-training**, BERT has already learned general language knowledge.  
In the **fine-tuning** phase, the pre-trained BERT model is adapted to perform a **specific NLP task** using a labeled dataset.

---

# Step 1: Load the Pre-trained BERT Model

Instead of training a model from scratch, we **load a pre-trained BERT model**.

The model already understands:

- Grammar
- Sentence structure
- Word meanings
- Contextual relationships

```text
Pre-trained BERT
        │
        ▼
Ready for Task-Specific Training
```

---

# Step 2: Add a Task-Specific Layer

A small output layer is added on top of BERT depending on the task.

Examples:

| NLP Task | Output Layer |
|----------|--------------|
| Sentiment Analysis | Classification Layer |
| Spam Detection | Classification Layer |
| Named Entity Recognition (NER) | Token Classification Layer |
| Question Answering | Start & End Prediction Layer |
| Text Similarity | Classification Layer |

---

# Step 3: Train on a Labeled Dataset

The entire model (or selected layers) is trained using labeled data.

```text
Labeled Dataset
        │
        ▼
Pre-trained BERT
        │
        ▼
Task-Specific Layer
        │
        ▼
Fine-Tuned Model
```

During fine-tuning:

- BERT adjusts its learned weights.
- The task-specific layer learns the target labels.
- The model becomes specialized for the new task.

---

# Transfer Learning Analogy

Fine-tuning is an example of **Transfer Learning**.

### Analogy

Imagine a student who has already completed a **Computer Science degree**.

- During college, the student learns programming, mathematics, algorithms, and problem-solving.
- Later, the student joins a company and receives training for a specific role, such as **Data Scientist** or **Machine Learning Engineer**.

Similarly:

- **Pre-training** = Learning general language knowledge.
- **Fine-tuning** = Learning a specific NLP task.

```text
Pre-training
      │
      ▼
General Language Knowledge
      │
      ▼
Fine-Tuning
      │
      ▼
Task-Specific Expertise
```

---

# Fine-Tuning Workflow

```text
Large Text Corpus
        │
        ▼
Pre-training
        │
        ▼
Pre-trained BERT Model
        │
        ▼
Load Pre-trained Model
        │
        ▼
Add Task-Specific Layer
        │
        ▼
Train on Labeled Dataset
        │
        ▼
Fine-Tuned BERT Model
```

---

# Advantages of Fine-Tuning

- Faster training than building a model from scratch.
- Requires less labeled data.
- Achieves higher accuracy.
- Leverages the knowledge learned during pre-training.
- Easily adaptable to different NLP tasks.

---

# Key Points

- Load a **pre-trained BERT model**.
- Add a **task-specific output layer**.
- Train using a **labeled dataset**.
- Fine-tuning is an application of **Transfer Learning**.
- Produces a model specialized for tasks such as:
  - Text Classification
  - Sentiment Analysis
  - Named Entity Recognition (NER)
  - Question Answering
  - Spam Detection

---

# One Pre-trained BERT → Multiple Fine-Tuned Models

```text
                 Pre-trained BERT
                       │
      ┌────────────────┼────────────────┐
      │                │                │
      ▼                ▼                ▼
Sentiment        Spam Detection      Question Answering
Analysis
      │
      ▼
Fine-Tuned Model 1

                 │
                 ▼
Named Entity Recognition (NER)
      │
      ▼
Fine-Tuned Model 2

                 │
                 ▼
Text Classification
      │
      ▼
Fine-Tuned Model 3

                 │
                 ▼
Machine Translation
      │
      ▼
Fine-Tuned Model 4

                 │
                 ▼
Text Summarization
      │
      ▼
Fine-Tuned Model 5
```

**Key Idea:**  
The **same pre-trained BERT model** serves as the foundation, but each task has its own **fine-tuned model** trained on task-specific labeled data.

---

# Three Phases of BERT

## Phase 1: Pre-training

- Train BERT on a **large unlabeled text corpus**.
- Learn general language understanding.
- Use two objectives:
  - **Masked Language Modeling (MLM)**
  - **Next Sentence Prediction (NSP)** *(used in the original BERT)*

**Output:** Pre-trained BERT model.

---

## Phase 2: Fine-Tuning

- Load the pre-trained BERT model.
- Add a task-specific output layer.
- Train on a labeled dataset for a specific NLP task.

Examples:

- Sentiment Analysis
- Text Classification
- Question Answering
- Named Entity Recognition (NER)
- Spam Detection

**Output:** Fine-tuned BERT model.

---

## Phase 3: Inference

Once the model is fine-tuned, it is used to make predictions on **new, unseen data**.

### Example

```text
Input:
"This movie was fantastic!"

        │
        ▼
Fine-Tuned Sentiment Model

        │
        ▼
Prediction:
Positive
```

During inference:

- No further training occurs.
- The model only performs prediction.
- Used in real-world applications such as chatbots, search engines, and recommendation systems.

---

# Complete BERT Workflow

```text
Large Unlabeled Text Corpus
            │
            ▼
      Phase 1: Pre-training
      (MLM + NSP)
            │
            ▼
     Pre-trained BERT Model
            │
            ▼
      Phase 2: Fine-Tuning
   (Task-Specific Dataset)
            │
            ▼
     Fine-Tuned BERT Model
            │
            ▼
      Phase 3: Inference
     (Predict New Data)
            │
            ▼
        Final Prediction
```

---

# Summary

| Phase | Purpose | Output |
|--------|---------|--------|
| **Phase 1: Pre-training** | Learn general language understanding from large unlabeled text | Pre-trained BERT |
| **Phase 2: Fine-Tuning** | Adapt BERT to a specific NLP task using labeled data | Fine-tuned BERT |
| **Phase 3: Inference** | Use the fine-tuned model to make predictions on new data | Predicted output |

---

# Key Points

- One **pre-trained BERT model** can produce **multiple fine-tuned models**.
- Each fine-tuned model is specialized for a different NLP task.
- The BERT pipeline consists of **three phases**:
  1. **Pre-training**
  2. **Fine-Tuning**
  3. **Inference**
- During **inference**, the model only predicts—it does not learn or update its weights.
